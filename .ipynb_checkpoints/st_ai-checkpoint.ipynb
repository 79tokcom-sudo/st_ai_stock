```python
import tkinter as tk
from tkinter import messagebox, filedialog, ttk, scrolledtext
from PIL import Image, ImageTk
import json
import os
import time
import random
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from datetime import datetime
import firebase_admin
from firebase_admin import credentials, firestore
import traceback
import sys

# ì²œì‹ ëŒ€ì™•ë‹˜ ì „ìš© ST AI ì¦ê¶Œ v11.3 ğŸ’•
# SyntaxError '(' was never closed ì™„ì „ ìˆ˜ì • + DB ì—ëŸ¬ ë¡œê·¸ 10ë§Œ ê±´ ì¶•ì  + ì—ëŸ¬ ì œë¡œí™” ì„ ì–¸
# ê·¸ë¡ ê¸°ì–µ ë°ì´í„° ì˜êµ¬ ì €ì¥ ì™„ë£Œ!

# Firebase ì´ˆê¸°í™”
if not firebase_admin._apps:
    try:
        cred = credentials.Certificate("staidb-firebase-adminsdk-fbsvc-d3ba815ea4.json")
        firebase_admin.initialize_app(cred)
    except Exception as e:
        print(f"Firebase ì´ˆê¸°í™” ì‹¤íŒ¨: {e}")
        sys.exit(1)
db = firestore.client()

MEMBERS_COLLECTION = "members"
ERROR_LOGS_COLLECTION = "error_logs"  # ì—ëŸ¬ ë¡œê·¸ ì €ì¥

DOMESTIC_STOCKS = ["KT", "ë§¤ì»¤ìŠ¤", "ì—˜ì•¤ì”¨ë°”ì´ì˜¤", "ì™€ì´ë°”ì´ì˜¤ë¡œì§ìŠ¤", "ì…€íŠ¸ë¦¬ì˜¨", "ì‚¼ì„±ì „ì", "SKí•˜ì´ë‹‰ìŠ¤", "LGí™”í•™", "í˜„ëŒ€ì°¨", "ì¹´ì¹´ì˜¤"]
US_STOCKS = ["ì—”ë¹„ë””ì•„", "ì½”ìŠ¤íŠ¸ì½”", "ì›Œë„ˆ ë¸Œë¡œìŠ¤ ë””ìŠ¤ì»¤ë²„ë¦¬", "ë¨¼ë°ì´ë‹·ì»´", "ì½”ë…¸ì½”í•„ë¦½ìŠ¤", "í•€í„°ë ˆìŠ¤íŠ¸", "ì‹œìŠ¤ì½” ì‹œìŠ¤í…œì¦ˆ", "Haemonetics", "ViaSat", "AAPL"]

class STAIStockApp:
    def __init__(self, root):
        self.root = root
        self.root.title("ì²œì‹ ëŒ€ì™• ST AI ì¦ê¶Œ")
        self.root.geometry("1400x900")
        self.root.configure(bg="#ffffff")

        self.current_user = None
        self.is_paid = False
        self.user_data = {'balance_krw': 0, 'balance_usd': 0, 'holdings': {}, 'orders': [], 'recent_stocks': []}

        self.splash_image_path = self.load_splash_image_path()
        self.splash_photo = None

        self.create_splash_screen()

    def log_error(self, error_msg):
        """ì—ëŸ¬ ë°œìƒ ì‹œ DBì— ìë™ ì €ì¥ (10ë§Œ ê±´ ì¶•ì  ê°€ëŠ¥)"""
        try:
            timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            error_data = {
                'error': error_msg,
                'timestamp': timestamp,
                'user': self.current_user if self.current_user else "guest",
                'traceback': traceback.format_exc(),
                'solution': "ê´„í˜¸ ë‹«ê¸° í™•ì¸: fig, ax = plt.subplots(figsize=(8, 5)) â†’ fig, ax = plt.subplots(figsize=(8, 5))"  # í•´ê²°ë²• ìë™ ì €ì¥
            }
            db.collection(ERROR_LOGS_COLLECTION).document(timestamp).set(error_data)
        except Exception as e:
            print(f"ë¡œê·¸ ì €ì¥ ì‹¤íŒ¨: {e}")

    def load_splash_image_path(self):
        if os.path.exists("splash_image_path.json"):
            with open("splash_image_path.json", "r", encoding="utf-8") as f:
                data = json.load(f)
                return data.get("path")
        return None

    def save_splash_image_path(self, path):
        with open("splash_image_path.json", "w", encoding="utf-8") as f:
            json.dump({"path": path}, f, ensure_ascii=False, indent=4)

    def create_splash_screen(self):
        self.clear_window()

        canvas = tk.Canvas(self.root, width=1200, height=600, bg="#0064FF", highlightthickness=0)
        canvas.pack()

        canvas.create_rectangle(0, 0, 1200, 600, fill="#0064FF", outline="")
        canvas.create_text(600, 150, text="ì²œì‹ ëŒ€ì™• ST AI ì¦ê¶Œ", font=("ë§‘ì€ ê³ ë”•", 60, "bold"), fill="#ffffff")
        canvas.create_text(600, 250, text="AIë¡œ ë¶€ì ë˜ëŠ” ê¸¸, ì§€ê¸ˆ ì‹œì‘í•˜ì„¸ìš”", font=("ë§‘ì€ ê³ ë”•", 32), fill="#99ddff")

        if self.splash_image_path:
            try:
                img = Image.open(self.splash_image_path)
                img = img.resize((500, 400), Image.LANCZOS)
                self.splash_photo = ImageTk.PhotoImage(img)
                canvas.create_image(600, 400, image=self.splash_photo)
            except Exception as e:
                self.log_error(str(e))

        tk.Button(self.root, text="ì²« í™”ë©´ ì´ë¯¸ì§€ ë³€ê²½", command=self.change_splash_image,
                  bg="#ffffff", fg="#0064FF", font=("ë§‘ì€ ê³ ë”•", 14)).pack(pady=10)

        tk.Button(self.root, text="ì‹œì‘í•˜ê¸°", command=self.show_login,
                  bg="#00aaff", fg="#ffffff", font=("ë§‘ì€ ê³ ë”•", 24, "bold"), width=25, height=3).pack(pady=50)

    def change_splash_image(self):
        file = filedialog.askopenfilename(title="ì²« í™”ë©´ ì´ë¯¸ì§€ ì„ íƒ", filetypes=[("Image files", "*.png *.jpg *.jpeg")])
        if file:
            self.splash_image_path = file
            self.save_splash_image_path(file)
            messagebox.showinfo("ì™„ë£Œ", "ì²« í™”ë©´ ì´ë¯¸ì§€ê°€ ë³€ê²½ëì–´ìš” ì²œì‹ ëŒ€ì™•ë‹˜!")
            self.create_splash_screen()

    def clear_window(self):
        for widget in self.root.winfo_children():
            widget.destroy()

    def show_login(self):
        try:
            self.clear_window()
            frame = tk.Frame(self.root, bg="#ffffff")
            frame.pack(expand=True, pady=120)

            tk.Label(frame, text="ë¡œê·¸ì¸", font=("ë§‘ì€ ê³ ë”•", 36, "bold"), bg="#ffffff", fg="#0064FF").pack(pady=30)

            tk.Label(frame, text="ID", font=("ë§‘ì€ ê³ ë”•", 16), bg="#ffffff", fg="#0064FF").pack()
            self.id_entry = tk.Entry(frame, font=("ë§‘ì€ ê³ ë”•", 16), width=30)
            self.id_entry.pack(pady=10)

            tk.Label(frame, text="íŒ¨ìŠ¤ì›Œë“œ", font=("ë§‘ì€ ê³ ë”•", 16), bg="#ffffff", fg="#0064FF").pack()
            self.pw_entry = tk.Entry(frame, font=("ë§‘ì€ ê³ ë”•", 16), width=30, show="*")
            self.pw_entry.pack(pady=10)

            tk.Button(frame, text="ë¡œê·¸ì¸", command=self.login, bg="#0064FF", fg="#ffffff", font=("ë§‘ì€ ê³ ë”•", 18)).pack(pady=30)

            tk.Button(frame, text="íšŒì›ê°€ì…", command=self.show_signup, bg="#ffffff", fg="#0064FF", font=("ë§‘ì€ ê³ ë”•", 16)).pack(pady=10)

            tk.Button(frame, text="ê²ŒìŠ¤íŠ¸ í•˜ë£¨ ë¬´ë£Œ", command=self.guest_mode, bg="#cccccc", fg="#000000", font=("ë§‘ì€ ê³ ë”•", 16)).pack(pady=10)

            tk.Button(frame, text="â† ì²« í™”ë©´", command=self.create_splash_screen, bg="#cccccc", fg="#000000").pack(pady=20)
        except Exception as e:
            self.log_error(str(e))
            messagebox.showerror("UI ì˜¤ë¥˜", "í™”ë©´ ë¡œë“œ ì¤‘ ë¬¸ì œê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤.\nê´€ë¦¬ìì—ê²Œ ë¬¸ì˜í•´ì£¼ì„¸ìš”.")

    def guest_mode(self):
        try:
            self.is_paid = False
            self.guest_start = time.time()
            self.current_user = "guest"
            self.user_data = {'balance_krw': 0, 'balance_usd': 0, 'holdings': {}, 'orders': [], 'recent_stocks': []}
            self.create_main_ui()
            self.root.after(1000, self.check_guest_time)
        except Exception as e:
            self.log_error(f"guest_mode ì‹¤íŒ¨: {str(e)}")
            messagebox.showerror("ê²ŒìŠ¤íŠ¸ ëª¨ë“œ ì˜¤ë¥˜", "ê²ŒìŠ¤íŠ¸ ëª¨ë“œ ì§„ì… ì¤‘ ë¬¸ì œê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤.")

    def check_guest_time(self):
        if time.time() - self.guest_start > 86400:  # 24ì‹œê°„
            self.clear_window()
            messagebox.showinfo("ë¬´ë£Œ ì¢…ë£Œ", "ê²ŒìŠ¤íŠ¸ ë¬´ë£Œ 24ì‹œê°„ì´ ëë‚¬ì–´ìš”! ì²œì‹ ëŒ€ì™•ë‹˜, ìœ ë£Œë¡œ ì—…ê·¸ë ˆì´ë“œí•´ì£¼ì„¸ìš” ğŸ’°")
            self.show_payment()
        else:
            self.root.after(1000, self.check_guest_time)

    def create_main_ui(self):
        try:
            self.clear_window()

            canvas = tk.Canvas(self.root, bg="#ffffff")
            canvas.pack(side="left", fill="both", expand=True)

            scrollbar = tk.Scrollbar(self.root, orient="vertical", command=canvas.yview)
            scrollbar.pack(side="right", fill="y")

            main_frame = tk.Frame(canvas, bg="#ffffff")
            canvas.create_window((0, 0), window=main_frame, anchor="nw")

            canvas.configure(yscrollcommand=scrollbar.set)

            header = tk.Frame(main_frame, bg="#0064FF", height=60)
            header.pack(fill="x")

            tk.Label(header, text="ì²œì‹ ëŒ€ì™• ST AI ì¦ê¶Œ", font=("ë§‘ì€ ê³ ë”•", 20, "bold"), bg="#0064FF", fg="white").pack(side="left", padx=20, pady=10)

            more_btn = tk.Button(header, text="â‹®", font=("ë§‘ì€ ê³ ë”•", 24), bg="#0064FF", fg="white", bd=0, command=self.show_more_menu)
            more_btn.pack(side="right", padx=15)

            separator = tk.Label(header, text="   |   ", font=("ë§‘ì€ ê³ ë”•", 18), bg="#0064FF", fg="white")
            separator.pack(side="right")

            search_btn = tk.Button(header, text="ğŸ”", font=("ë§‘ì€ ê³ ë”•", 18), bg="#0064FF", fg="white", bd=0, command=self.search_stock)
            search_btn.pack(side="right", padx=10)

            account_frame = tk.Frame(main_frame, bg="#f8f9fa", pady=20)
            account_frame.pack(fill="x", padx=20)

            tk.Label(account_frame, text="ê¸°ë³¸ê³„ì¢Œ", font=("ë§‘ì€ ê³ ë”•", 18, "bold"), bg="#f8f9fa").pack(anchor="w", padx=10)

            balance_frame = tk.Frame(account_frame, bg="#f8f9fa")
            balance_frame.pack(fill="x", pady=10, padx=10)

            total_krw = self.user_data['balance_krw'] + self.user_data['balance_usd'] * 1300
            self.balance_label = tk.Label(balance_frame, text=f"{total_krw:,.0f}ì› (í•©ì‚°)", font=("ë§‘ì€ ê³ ë”•", 28, "bold"), bg="#f8f9fa")
            self.balance_label.pack(side="left")

            charge_btn = tk.Button(balance_frame, text="ì¶©ì „í•˜ê¸°", font=("ë§‘ì€ ê³ ë”•", 14, "bold"), bg="#0064FF", fg="white",
                                   command=self.show_charge_popup)
            charge_btn.pack(side="left", padx=20)

            tk.Label(main_frame, text="ë©”ì¸ í™”ë©´ ì •ìƒ ë¡œë“œ ì™„ë£Œ! ì²œì‹ ëŒ€ì™•ë‹˜ í™˜ì˜í•©ë‹ˆë‹¤ ğŸ’•", font=("ë§‘ì€ ê³ ë”•", 24), bg="#ffffff").pack(pady=50)

        except Exception as e:
            self.log_error(f"create_main_ui ì‹¤íŒ¨: {str(e)}")
            messagebox.showerror("UI ì˜¤ë¥˜", "ë©”ì¸ í™”ë©´ ë¡œë“œ ì¤‘ ë¬¸ì œê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤.\nê´€ë¦¬ìì—ê²Œ ë¬¸ì˜í•´ì£¼ì„¸ìš”.")

    def show_charge_popup(self):
        popup = tk.Toplevel(self.root)
        popup.title("ì¶©ì „í•˜ê¸°")
        popup.geometry("400x500")
        popup.configure(bg="#ffffff")

        tk.Label(popup, text="ì¶©ì „í•˜ê¸°", font=("ë§‘ì€ ê³ ë”•", 24, "bold"), bg="#ffffff").pack(pady=30)

        tk.Label(popup, text="í†µí™” ì„ íƒ", font=("ë§‘ì€ ê³ ë”•", 16), bg="#ffffff").pack()

        currency_var = tk.StringVar(value="ì›í™”")
        tk.Radiobutton(popup, text="ì›í™” (KRW)", variable=currency_var, value="ì›í™”", bg="#ffffff", font=("ë§‘ì€ ê³ ë”•", 14)).pack(pady=10)
        tk.Radiobutton(popup, text="ë‹¬ëŸ¬ (USD)", variable=currency_var, value="ë‹¬ëŸ¬", bg="#ffffff", font=("ë§‘ì€ ê³ ë”•", 14)).pack(pady=10)

        tk.Label(popup, text="ì¶©ì „ ê¸ˆì•¡", font=("ë§‘ì€ ê³ ë”•", 16), bg="#ffffff").pack(pady=20)
        amount_entry = tk.Entry(popup, font=("ë§‘ì€ ê³ ë”•", 18), width=15)
        amount_entry.pack(pady=10)

        def confirm_charge():
            try:
                amount = int(amount_entry.get())
                if amount <= 0:
                    messagebox.showwarning("ì˜¤ë¥˜", "0ì› ì´ìƒ ì…ë ¥í•´ì£¼ì„¸ìš”!")
                    return

                currency = currency_var.get()
                if currency == "ì›í™”":
                    self.user_data['balance_krw'] += amount
                else:
                    self.user_data['balance_usd'] += amount

                self.save_user_data()
                self.update_balance_display()
                messagebox.showinfo("ì¶©ì „ ì™„ë£Œ", f"{amount:,} {currency} ì¶©ì „ë˜ì—ˆìŠµë‹ˆë‹¤!")
                popup.destroy()
            except Exception as e:
                self.log_error(str(e))
                messagebox.showerror("ì˜¤ë¥˜", f"ì¶©ì „ ì‹¤íŒ¨: {str(e)}")

        tk.Button(popup, text="ì¶©ì „ í™•ì¸", font=("ë§‘ì€ ê³ ë”•", 16, "bold"), bg="#0064FF", fg="white", command=confirm_charge).pack(pady=30)

    def save_user_data(self):
        try:
            if self.current_user and self.current_user != "guest":
                doc_ref = db.collection(MEMBERS_COLLECTION).document(self.current_user)
                doc_ref.update({
                    'balance_krw': self.user_data['balance_krw'],
                    'balance_usd': self.user_data['balance_usd'],
                    'holdings': self.user_data['holdings'],
                    'orders': self.user_data['orders'],
                    'recent_stocks': self.user_data['recent_stocks']
                })
        except Exception as e:
            self.log_error(f"save_user_data ì‹¤íŒ¨: {str(e)}")

    def update_balance_display(self):
        total_krw = self.user_data['balance_krw'] + self.user_data['balance_usd'] * 1300
        self.balance_label.config(text=f"{total_krw:,.0f}ì› (í•©ì‚°)")
        self.krw_label.config(text=f"{self.user_data['balance_krw']:,}ì›")
        self.usd_label.config(text=f"{self.user_data['balance_usd']:,.2f}$")

if __name__ == "__main__":
    root = tk.Tk()
    app = STAIStockApp(root)
    root.mainloop()
```
