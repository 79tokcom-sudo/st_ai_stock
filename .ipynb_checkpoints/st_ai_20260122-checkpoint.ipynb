```python
# -*- coding: utf-8 -*-
"""
ì²œì‹ ëŒ€ì™• ST AI ì¦ê¶Œ (ONE UI) + Voice AI í†µí•© ì›ì†ŒìŠ¤ ì˜ˆì‹œ
- Tkinter UI
- Firestore(ì„ íƒ) 1íšŒ ì—°ê²° + batch ì €ì¥
- Voice: ë…¹ìŒ -> STT -> ëª…ë ¹ íŒŒì‹± -> ì•± ê¸°ëŠ¥ ì‹¤í–‰ -> TTS ì‘ë‹µ
- í‚¤ëŠ” OPENAI_API_KEY í™˜ê²½ë³€ìˆ˜ë¡œë§Œ ì½ìŒ (ì½”ë“œ/DB ì €ì¥ ê¸ˆì§€)

í•„ìˆ˜ ì„¤ì¹˜:
  pip install sounddevice soundfile numpy requests
ê¶Œì¥(í™•ì‹¤í•œ mp3 ì¬ìƒ):
  pip install playsound==1.2.2

(ì„ íƒ) Firestore:
  pip install firebase-admin google-cloud-firestore
"""

import os
import re
import json
import time
import queue
import threading
import tempfile
from dataclasses import dataclass
from typing import Optional, Tuple, Dict, Any, List

import numpy as np
import requests
import sounddevice as sd
import soundfile as sf

# playsound(ê¶Œì¥)
try:
    from playsound import playsound
    HAS_PLAYSOUND = True
except Exception:
    HAS_PLAYSOUND = False

# (ì„ íƒ) Firestore
HAS_FIRESTORE = False
try:
    import firebase_admin
    from firebase_admin import credentials
    from firebase_admin import firestore
    HAS_FIRESTORE = True
except Exception:
    HAS_FIRESTORE = False

# -------------------------
# 0) ì „ì—­ ì„¤ì •(ë³´ì•ˆ/ì›ì¹™)
# -------------------------
APP_VERSION = "v18.3-voice-merge"
ONE_UI_NAME = "ì²œì‹ ëŒ€ì™• ST AI ì¦ê¶Œ (ONE UI)"
KST_TZ = "Asia/Seoul"

# í™˜ê²½ë³€ìˆ˜ ì´ë¦„ (í‚¤ ê°’ ì ˆëŒ€ í•˜ë“œì½”ë”© ê¸ˆì§€)
OPENAI_API_KEY_ENV = "OPENAI_API_KEY"

# OpenAI ìŒì„± API (STT/TTS)
OPENAI_STT_URL = "https://api.openai.com/v1/audio/transcriptions"
OPENAI_TTS_URL = "https://api.openai.com/v1/audio/speech"
OPENAI_STT_MODEL = "gpt-4o-mini-transcribe"  # ì•ˆë˜ë©´ whisper-1ë¡œ êµì²´
OPENAI_TTS_MODEL = "gpt-4o-mini-tts"         # ì•ˆë˜ë©´ tts-1ë¡œ êµì²´
OPENAI_TTS_VOICE = "alloy"
OPENAI_TTS_FORMAT = "mp3"

# ìŒì„± ë…¹ìŒ
SAMPLE_RATE = 16000
CHANNELS = 1
RECORD_SECONDS = 6

# Firestore JSON ê²½ë¡œ (ì‚¬ìš©ì ì œê³µ íŒŒì¼ëª… ê¸°ì¤€)
FIREBASE_CRED_PATH = "staidb-firebase-adminsdk-fbsvc-d3ba815ea4.json"

# DB 1MB ì œí•œ ëŒ€ë¹„ ë¶„í• (ë¬¸ìì—´ ì €ì¥ìš©)
MAX_BYTES_TEXT = 900_000


# -------------------------
# 1) ìœ í‹¸
# -------------------------
def now_ts() -> float:
    return time.time()

def split_text_by_bytes(text: str, max_bytes: int = MAX_BYTES_TEXT) -> List[str]:
    if not text:
        return []
    chunks = []
    cur = ""
    cur_bytes = 0
    for ch in text:
        b = ch.encode("utf-8")
        if cur_bytes + len(b) > max_bytes:
            chunks.append(cur)
            cur = ch
            cur_bytes = len(b)
        else:
            cur += ch
            cur_bytes += len(b)
    if cur:
        chunks.append(cur)
    return chunks

def safe_env(key: str) -> str:
    v = os.getenv(key, "").strip()
    return v

def get_openai_key() -> str:
    key = safe_env(OPENAI_API_KEY_ENV)
    if not key:
        raise RuntimeError(
            f"OpenAI API Keyê°€ ì—†ìŠµë‹ˆë‹¤. í™˜ê²½ë³€ìˆ˜ {OPENAI_API_KEY_ENV} ë¥¼ ì„¤ì •í•˜ì„¸ìš”.\n"
            f"Windows ì˜ˆ: setx {OPENAI_API_KEY_ENV} \"í‚¤ê°’\"  (ì„¤ì • í›„ ìƒˆ CMD/IDEì—ì„œ ì‹¤í–‰)"
        )
    return key


# -------------------------
# 2) Firestore: 1íšŒ init + batch ì €ì¥ ê³¨ê²©
# -------------------------
class FirestoreClient:
    def __init__(self, enable: bool = True):
        self.enable = enable and HAS_FIRESTORE
        self.db = None

    def init_once(self):
        """ì—°ê²°ì€ í•œ ë²ˆë§Œ"""
        if not self.enable:
            return
        if self.db is not None:
            return
        if not os.path.exists(FIREBASE_CRED_PATH):
            raise RuntimeError(f"Firestore ìê²©ì¦ëª… íŒŒì¼ ì—†ìŒ: {FIREBASE_CRED_PATH}")
        try:
            # ì´ë¯¸ init ë˜ì—ˆìœ¼ë©´ ì˜ˆì™¸ê°€ ë‚  ìˆ˜ ìˆìœ¼ë¯€ë¡œ ë°©ì–´
            if not firebase_admin._apps:
                cred = credentials.Certificate(FIREBASE_CRED_PATH)
                firebase_admin.initialize_app(cred)
            self.db = firestore.client()
        except Exception as e:
            raise RuntimeError(f"Firestore init ì‹¤íŒ¨: {e}")

    def batch_write(self, ops: List[Tuple[str, str, Dict[str, Any]]]):
        """
        ops: [(collection, doc_id, data), ...]
        - í•œë²ˆì— ì²˜ë¦¬ (ì—°ê²° ìì£¼ í•˜ì§€ ì•Šê¸°)
        """
        if not self.enable:
            return
        self.init_once()
        batch = self.db.batch()
        for col, doc_id, data in ops:
            ref = self.db.collection(col).document(doc_id)
            batch.set(ref, data, merge=True)
        batch.commit()


# -------------------------
# 3) OpenAI STT/TTS
# -------------------------
def openai_stt_transcribe(wav_path: str) -> str:
    api_key = get_openai_key()
    headers = {"Authorization": f"Bearer {api_key}"}
    with open(wav_path, "rb") as f:
        files = {"file": (os.path.basename(wav_path), f, "audio/wav")}
        data = {"model": OPENAI_STT_MODEL}
        r = requests.post(OPENAI_STT_URL, headers=headers, files=files, data=data, timeout=90)
    if r.status_code != 200:
        raise RuntimeError(f"STT ì‹¤íŒ¨: {r.status_code} / {r.text}")
    j = r.json()
    return (j.get("text") or "").strip()

def openai_tts_bytes(text: str) -> bytes:
    api_key = get_openai_key()
    headers = {
        "Authorization": f"Bearer {api_key}",
        "Content-Type": "application/json",
    }
    payload = {
        "model": OPENAI_TTS_MODEL,
        "voice": OPENAI_TTS_VOICE,
        "format": OPENAI_TTS_FORMAT,
        "input": text,
    }
    r = requests.post(OPENAI_TTS_URL, headers=headers, data=json.dumps(payload), timeout=90)
    if r.status_code != 200:
        raise RuntimeError(f"TTS ì‹¤íŒ¨: {r.status_code} / {r.text}")
    return r.content

def play_mp3_bytes(mp3_data: bytes):
    tmp = tempfile.NamedTemporaryFile(delete=False, suffix=".mp3")
    tmp.write(mp3_data)
    tmp.close()
    path = tmp.name

    # playsoundê°€ ê°€ì¥ í™•ì‹¤
    if HAS_PLAYSOUND:
        playsound(path)
        return

    # fallback: ìœˆë„ìš° ê¸°ë³¸ ì‹¤í–‰(í”Œë ˆì´ì–´ ì—´ë¦¼)
    try:
        if os.name == "nt":
            os.startfile(path)  # noqa
        else:
            import subprocess
            subprocess.Popen(["xdg-open", path])
    except Exception:
        pass

def speak_async(text: str, log_fn):
    def _run():
        try:
            mp3 = openai_tts_bytes(text)
            play_mp3_bytes(mp3)
        except Exception as e:
            log_fn(f"[TTS ERROR] {e}")
    threading.Thread(target=_run, daemon=True).start()


# -------------------------
# 4) ë…¹ìŒ
# -------------------------
def record_wav(seconds: int = RECORD_SECONDS, sample_rate: int = SAMPLE_RATE, channels: int = CHANNELS) -> str:
    audio = sd.rec(int(seconds * sample_rate), samplerate=sample_rate, channels=channels, dtype="float32")
    sd.wait()

    audio_int16 = np.clip(audio, -1.0, 1.0)
    audio_int16 = (audio_int16 * 32767).astype(np.int16)

    wav_path = tempfile.NamedTemporaryFile(delete=False, suffix=".wav").name
    sf.write(wav_path, audio_int16, sample_rate, subtype="PCM_16")
    return wav_path


# -------------------------
# 5) ëª…ë ¹ íŒŒì‹±(ì˜ˆì‹œ): "ì‚¼ì„±ì „ì 10ë§Œì› ë§¤ìˆ˜" ê°™ì€ ì…ë ¥ì„ êµ¬ì¡°í™”
# -------------------------
@dataclass
class ParsedCommand:
    intent: str
    symbol: Optional[str] = None
    amount_krw: Optional[int] = None
    raw: str = ""

def parse_korean_money_to_int(s: str) -> Optional[int]:
    # "10ë§Œì›" "5ë§Œ" "100000" ë“± ë‹¨ìˆœ ì§€ì›
    s = s.replace(",", "").strip()
    m = re.search(r"(\d+)\s*ë§Œ\s*ì›", s)
    if m:
        return int(m.group(1)) * 10_000
    m = re.search(r"(\d+)\s*ë§Œ", s)
    if m:
        return int(m.group(1)) * 10_000
    m = re.search(r"(\d+)\s*ì›", s)
    if m:
        return int(m.group(1))
    m = re.search(r"\b(\d{4,})\b", s)
    if m:
        return int(m.group(1))
    return None

def parse_voice_command(text: str) -> ParsedCommand:
    t = text.strip()
    low = t.replace(" ", "")

    if "í˜„ì¬ì”ê³ " in low or (("ì”ê³ " in low) and ("ì¶©ì „" not in low)):
        return ParsedCommand(intent="OPEN_BALANCE", raw=t)

    if "ì„¤ì •" in low:
        return ParsedCommand(intent="OPEN_SETTINGS", raw=t)

    if "ë”ë³´ê¸°" in low:
        return ParsedCommand(intent="OPEN_MORE", raw=t)

    if "ì£¼ë¬¸ë‚´ì—­" in low or "ì£¼ë¬¸" in low:
        return ParsedCommand(intent="OPEN_ORDERS", raw=t)

    if "ì¶”ì²œ" in low and ("ì¢…ëª©" in low or "ì¶”ì²œì¢…ëª©" in low):
        return ParsedCommand(intent="OPEN_RECOMMEND", raw=t)

    # ë§¤ìˆ˜/ë§¤ë„
    if "ë§¤ìˆ˜" in low:
        # ì•„ì£¼ ë‹¨ìˆœ: "ì‚¼ì„±ì „ì 10ë§Œì› ë§¤ìˆ˜"
        amount = parse_korean_money_to_int(t)
        # ì‹¬ë³¼ì€ ë¬¸ìì—´ì—ì„œ ê¸ˆì•¡/ë§¤ìˆ˜ ì œê±°í•˜ê³  ë‚¨ì€ ë‹¨ì–´ë¥¼ ì‹¬ë³¼ë¡œ ê°„ì£¼(ì˜ˆì‹œ)
        symbol = re.sub(r"\d+|ë§Œ\s*ì›|ë§Œì›|ì›|ë§¤ìˆ˜| ", "", t).strip()
        symbol = symbol if symbol else None
        return ParsedCommand(intent="BUY", symbol=symbol, amount_krw=amount, raw=t)

    if "ë§¤ë„" in low:
        amount = parse_korean_money_to_int(t)
        symbol = re.sub(r"\d+|ë§Œ\s*ì›|ë§Œì›|ì›|ë§¤ë„| ", "", t).strip()
        symbol = symbol if symbol else None
        return ParsedCommand(intent="SELL", symbol=symbol, amount_krw=amount, raw=t)

    return ParsedCommand(intent="UNKNOWN", raw=t)


# -------------------------
# 6) ST AI ì¦ê¶Œ ì•± ê³¨ê²©(ì—¬ê¸°ì— ë‹¹ì‹  v18.3 ê¸°ëŠ¥ì„ ì—°ê²°)
# -------------------------
class STApp:
    """
    ì—¬ê¸° í´ë˜ìŠ¤/í•¨ìˆ˜ë“¤ì„ 'ë‹¹ì‹  v18.3 ì›ë³¸ ì½”ë“œ'ì— ë§ì¶° ì—°ê²°í•˜ë©´ ë©ë‹ˆë‹¤.
    UIëŠ” ì‚¬ìš©ìê°€ ìš”ì²­í•  ë•Œë§Œ ë³€ê²½í•œë‹¤ëŠ” ì›ì¹™ì„ ì§€í‚¤ê¸° ìœ„í•´
    ì•„ë˜ëŠ” 'ì¶”ê°€ UI ìµœì†Œí™”'ë¡œ ìŒì„±ë²„íŠ¼ë§Œ ìƒë‹¨ì— ë¶™ì…ë‹ˆë‹¤.
    """
    def __init__(self, root):
        self.root = root
        self.db = FirestoreClient(enable=True)  # í•„ìš” ì—†ìœ¼ë©´ False
        self.member_id = "guest"                # TODO: ì‹¤ì œ ë¡œê·¸ì¸ member_id ì—°ê²°
        self.cash_balance = 0                   # TODO: ì‹¤ì œ ì”ê³  ì—°ê²°
        self.log_lines: List[str] = []

        # ---- UI (ONE UI ìµœì†Œ ê³¨ê²©) ----
        import tkinter as tk
        self.tk = tk
        root.title(f"{ONE_UI_NAME} {APP_VERSION}")
        root.geometry("1100x700")

        # ìƒë‹¨ ë°”
        self.top = tk.Frame(root)
        self.top.pack(side="top", fill="x")

        self.lbl = tk.Label(self.top, text=f"{ONE_UI_NAME}  |  {APP_VERSION}", anchor="w")
        self.lbl.pack(side="left", padx=8, pady=6)

        # ìŒì„± ë²„íŠ¼ ì˜ì—­(ìƒë‹¨ ì˜¤ë¥¸ìª½)
        self.voice_frame = tk.Frame(self.top)
        self.voice_frame.pack(side="right", padx=8)

        self.btn_voice = tk.Button(self.voice_frame, text="ğŸ¤ ìŒì„± ëª…ë ¹", command=self.on_voice_button)
        self.btn_voice.pack(side="left", padx=4)

        self.voice_status = tk.Label(self.voice_frame, text="ëŒ€ê¸°ì¤‘")
        self.voice_status.pack(side="left", padx=6)

        # ë©”ì¸ ì˜ì—­
        self.main = tk.Frame(root)
        self.main.pack(side="top", fill="both", expand=True)

        # ê°„ë‹¨ ë¡œê·¸ ë°•ìŠ¤
        self.txt = tk.Text(self.main, height=20)
        self.txt.pack(side="bottom", fill="x")

        self._voice_q = queue.Queue()
        self._voice_busy = False

        # í í´ë§
        self.root.after(120, self._poll_voice_queue)

        # Firestore 1íšŒ ì´ˆê¸°í™”(ì›ì¹™ìƒ ì•± ì‹œì‘ ì‹œ 1ë²ˆë§Œ)
        # ì‹¤íŒ¨í•´ë„ ì•±ì´ ì£½ì§€ ì•Šê²Œ try
        try:
            self.db.init_once()
            self.log("[DB] Firestore init ok")
        except Exception as e:
            self.log(f"[DB] Firestore disabled/fail: {e}")

    # ---------------- UI/ë¡œê·¸ ----------------
    def log(self, msg: str):
        ts = time.strftime("%H:%M:%S")
        line = f"[{ts}] {msg}"
        self.log_lines.append(line)
        try:
            self.txt.insert("end", line + "\n")
            self.txt.see("end")
        except Exception:
            pass

    def set_voice_status(self, s: str):
        try:
            self.voice_status.config(text=s)
        except Exception:
            pass

    def popup_error(self, title: str, msg: str):
        try:
            from tkinter import messagebox
            messagebox.showerror(title, msg)
        except Exception:
            self.log(f"[POPUP FAIL] {title}: {msg}")

    # ---------------- ST ê¸°ëŠ¥(ì—°ê²° ì§€ì ) ----------------
    def open_balance_screen(self):
        # TODO: v18.3 "í˜„ì¬ì”ê³ " í™”ë©´/íƒ­ ì—´ê¸° í•¨ìˆ˜ë¡œ ì—°ê²°
        self.log("âœ… í˜„ì¬ì”ê³  í™”ë©´(ì—°ê²° í•„ìš”)")

    def open_settings_screen(self):
        # TODO: v18.3 ì„¤ì • ì—´ê¸°
        self.log("âœ… ì„¤ì • í™”ë©´(ì—°ê²° í•„ìš”)")

    def open_more_screen(self):
        # TODO: v18.3 ë”ë³´ê¸° ì—´ê¸°(1x2 ë ˆì´ì•„ì›ƒ ìœ ì§€)
        self.log("âœ… ë”ë³´ê¸° í™”ë©´(ì—°ê²° í•„ìš”)")

    def open_orders_screen(self):
        # TODO: v18.3 ì£¼ë¬¸ë‚´ì—­ ì—´ê¸°
        self.log("âœ… ì£¼ë¬¸ë‚´ì—­ í™”ë©´(ì—°ê²° í•„ìš”)")

    def open_recommend_screen(self):
        # TODO: v18.3 ì¶”ì²œì¢…ëª© íƒ­ ì—´ê¸°
        self.log("âœ… ì¶”ì²œì¢…ëª© í™”ë©´(ì—°ê²° í•„ìš”)")

    def place_buy_order(self, symbol: Optional[str], amount_krw: Optional[int]):
        # TODO: v18.3 ì£¼ì‹ì˜¤ë” ìƒì„± ë¡œì§ + ì”ê³  ì²´í¬ + ì‹œì¥ê°€ readonly ê·œì¹™ ì¤€ìˆ˜
        self.log(f"âœ… ë§¤ìˆ˜ ìš”ì²­(ì—°ê²° í•„ìš”) symbol={symbol} amount={amount_krw}")

    def place_sell_order(self, symbol: Optional[str], amount_krw: Optional[int]):
        # TODO: v18.3 ë§¤ë„ ë¡œì§
        self.log(f"âœ… ë§¤ë„ ìš”ì²­(ì—°ê²° í•„ìš”) symbol={symbol} amount={amount_krw}")

    # ---------------- DB ê¸°ë¡(ìŒì„±ë¡œê·¸/ì—ëŸ¬ë¡œê·¸) ----------------
    def save_voice_log_batch(self, raw_text: str, intent: str, result_text: str, error_text: Optional[str] = None):
        """
        DB 1MB ë„˜ì§€ ì•Šê²Œ ë¶„í•  ì €ì¥ + batch
        - ì—°ê²° ìì£¼ í•˜ì§€ ë§ê¸°: ì—¬ê¸°ì„œ opsë§Œ ë§Œë“¤ì–´ í•œë²ˆì— commit
        """
        try:
            rid = f"voice_{int(now_ts()*1000)}"
            parts_raw = split_text_by_bytes(raw_text)
            parts_result = split_text_by_bytes(result_text)
            parts_error = split_text_by_bytes(error_text) if error_text else []

            data = {
                "member_id": self.member_id,
                "app_version": APP_VERSION,
                "created_at": firestore.SERVER_TIMESTAMP if (HAS_FIRESTORE and self.db.db) else time.time(),
                "intent": intent,
                "raw_parts": parts_raw,
                "result_parts": parts_result,
                "error_parts": parts_error,
            }

            # í•„ìš”í•˜ë‹¤ë©´ programs/UI ë²„ì „ ê¸°ë¡ë„ ì—¬ê¸°ì— opsë¡œ ê°™ì´ ë¬¶ì–´ì„œ ì €ì¥ ê°€ëŠ¥
            ops = [("voice_logs", rid, data)]
            self.db.batch_write(ops)
            self.log(f"[DB] voice log saved: {rid}")
        except Exception as e:
            # DB ì €ì¥ ì‹¤íŒ¨ëŠ” ì•± ê¸°ëŠ¥ì„ ë§‰ì§€ ì•ŠìŒ
            self.log(f"[DB ERROR] save_voice_log_batch: {e}")

    # ---------------- Voice Flow ----------------
    def on_voice_button(self):
        if self._voice_busy:
            return
        self._voice_busy = True
        self.btn_voice.config(state="disabled")
        self.set_voice_status("ğŸ™ï¸ ë…¹ìŒì¤‘...")

        th = threading.Thread(target=self._voice_worker, daemon=True)
        th.start()

    def _voice_worker(self):
        raw_text = ""
        intent = "UNKNOWN"
        result_text = ""
        err_text = None

        try:
            # 1) ë…¹ìŒ
            wav = record_wav(RECORD_SECONDS, SAMPLE_RATE, CHANNELS)
            self._voice_q.put(("status", "ğŸ§  ìŒì„± ì¸ì‹ì¤‘..."))

            # 2) STT
            raw_text = openai_stt_transcribe(wav)
            if not raw_text:
                result_text = "ìŒì„±ì´ ì¸ì‹ë˜ì§€ ì•Šì•˜ì–´ìš”. ë˜ë ·í•˜ê²Œ ë‹¤ì‹œ ë§í•´ì¤˜ìš”."
                self._voice_q.put(("done_result", raw_text, "NO_SPEECH", result_text, None))
                return

            # 3) ëª…ë ¹ íŒŒì‹±
            cmd = parse_voice_command(raw_text)
            intent = cmd.intent
            self._voice_q.put(("status", f"âš™ï¸ ëª…ë ¹ ì‹¤í–‰ì¤‘({intent})..."))

            # 4) ì•± ê¸°ëŠ¥ ì‹¤í–‰ (UI ì‘ì—…ì€ ë©”ì¸ì“°ë ˆë“œì—ì„œ)
            #    -> ì—¬ê¸°ì„œëŠ” ì‹¤í–‰ ìš”ì²­ë§Œ íë¡œ ë³´ëƒ„
            self._voice_q.put(("execute", cmd))

            # execute ê²°ê³¼ëŠ” ë©”ì¸ì—ì„œ ì²˜ë¦¬í•˜ê³  TTSê¹Œì§€ ìˆ˜í–‰
            return

        except Exception as e:
            err_text = str(e)
            result_text = f"ìŒì„± ì²˜ë¦¬ ì¤‘ ì˜¤ë¥˜ê°€ ë°œìƒí–ˆì–´ìš”: {err_text}"
            self._voice_q.put(("error", raw_text, intent, result_text, err_text))

        finally:
            # busy í•´ì œëŠ” pollì—ì„œ ì²˜ë¦¬

            pass

    def _poll_voice_queue(self):
        try:
            while True:
                item = self._voice_q.get_nowait()
                kind = item[0]

                if kind == "status":
                    self.set_voice_status(item[1])

                elif kind == "execute":
                    cmd: ParsedCommand = item[1]
                    # ---- ì—¬ê¸°ì„œ ì‹¤ì œ UI/ê¸°ëŠ¥ ì‹¤í–‰ ----
                    result_text, say_text, err_text = self._execute_parsed_command(cmd)

                    # DB ì €ì¥(í•œ ë²ˆì—)
                    self.save_voice_log_batch(cmd.raw, cmd.intent, result_text, err_text)

                    # TTS (ë¹„ë™ê¸°)
                    if say_text:
                        speak_async(say_text, self.log)

                    self.set_voice_status("âœ… ì™„ë£Œ")
                    self._voice_finish()

                elif kind == "done_result":
                    raw_text, intent, result_text, err_text = item[1], item[2], item[3], item[4]
                    self.save_voice_log_batch(raw_text, intent, result_text, err_text)
                    speak_async(result_text, self.log)
                    self.set_voice_status("âœ… ì™„ë£Œ")
                    self._voice_finish()

                elif kind == "error":
                    raw_text, intent, result_text, err_text = item[1], item[2], item[3], item[4]
                    self.log(f"[VOICE ERROR] {err_text}")
                    self.save_voice_log_batch(raw_text, intent, result_text, err_text)
                    self.popup_error("ìŒì„± AI ì˜¤ë¥˜", err_text or "Unknown error")
                    self.set_voice_status("âŒ ì˜¤ë¥˜")
                    self._voice_finish()

        except queue.Empty:
            pass

        self.root.after(120, self._poll_voice_queue)

    def _voice_finish(self):
        self._voice_busy = False
        try:
            self.btn_voice.config(state="normal")
        except Exception:
            pass
        self.set_voice_status("ëŒ€ê¸°ì¤‘")

    def _execute_parsed_command(self, cmd: ParsedCommand) -> Tuple[str, str, Optional[str]]:
        """
        return: (result_text, tts_text, error_text)
        """
        try:
            if cmd.intent == "OPEN_BALANCE":
                self.open_balance_screen()
                return "í˜„ì¬ì”ê³  í™”ë©´ì„ ì—´ì—ˆìŠµë‹ˆë‹¤.", "í˜„ì¬ì”ê³  í™”ë©´ì„ ì—´ì—ˆìŠµë‹ˆë‹¤.", None

            if cmd.intent == "OPEN_SETTINGS":
                self.open_settings_screen()
                return "ì„¤ì • í™”ë©´ì„ ì—´ì—ˆìŠµë‹ˆë‹¤.", "ì„¤ì • í™”ë©´ì„ ì—´ì—ˆìŠµë‹ˆë‹¤.", None

            if cmd.intent == "OPEN_MORE":
                self.open_more_screen()
                return "ë”ë³´ê¸° í™”ë©´ì„ ì—´ì—ˆìŠµë‹ˆë‹¤.", "ë”ë³´ê¸° í™”ë©´ì„ ì—´ì—ˆìŠµë‹ˆë‹¤.", None

            if cmd.intent == "OPEN_ORDERS":
                self.open_orders_screen()
                return "ì£¼ë¬¸ë‚´ì—­ í™”ë©´ì„ ì—´ì—ˆìŠµë‹ˆë‹¤.", "ì£¼ë¬¸ë‚´ì—­ í™”ë©´ì„ ì—´ì—ˆìŠµë‹ˆë‹¤.", None

            if cmd.intent == "OPEN_RECOMMEND":
                self.open_recommend_screen()
                return "ì¶”ì²œì¢…ëª© í™”ë©´ì„ ì—´ì—ˆìŠµë‹ˆë‹¤.", "ì¶”ì²œì¢…ëª© í™”ë©´ì„ ì—´ì—ˆìŠµë‹ˆë‹¤.", None

            if cmd.intent == "BUY":
                self.place_buy_order(cmd.symbol, cmd.amount_krw)
                say = "ë§¤ìˆ˜ ëª…ë ¹ì„ í™•ì¸í–ˆì–´ìš”."
                return f"ë§¤ìˆ˜ ëª…ë ¹ í™•ì¸: ì¢…ëª©={cmd.symbol}, ê¸ˆì•¡={cmd.amount_krw}", say, None

            if cmd.intent == "SELL":
                self.place_sell_order(cmd.symbol, cmd.amount_krw)
                say = "ë§¤ë„ ëª…ë ¹ì„ í™•ì¸í–ˆì–´ìš”."
                return f"ë§¤ë„ ëª…ë ¹ í™•ì¸: ì¢…ëª©={cmd.symbol}, ê¸ˆì•¡={cmd.amount_krw}", say, None

            # UNKNOWN
            msg = ("ì§€ì›í•˜ì§€ ì•ŠëŠ” ëª…ë ¹ì…ë‹ˆë‹¤.\n"
                   "ì˜ˆ: 'í˜„ì¬ì”ê³  ë³´ì—¬ì¤˜', 'ì„¤ì • ë“¤ì–´ê°€', 'ë”ë³´ê¸° ì—´ì–´', 'ì£¼ë¬¸ë‚´ì—­ ë³´ì—¬ì¤˜', 'ì‚¼ì„±ì „ì 10ë§Œì› ë§¤ìˆ˜'")
            return msg, "ì§€ì›í•˜ì§€ ì•ŠëŠ” ëª…ë ¹ì´ì—ìš”.", None

        except Exception as e:
            err = str(e)
            return f"ëª…ë ¹ ì‹¤í–‰ ì‹¤íŒ¨: {err}", "ëª…ë ¹ ì‹¤í–‰ ì¤‘ ì˜¤ë¥˜ê°€ ë°œìƒí–ˆì–´ìš”.", err


# -------------------------
# 7) ë©”ì¸ ì‹¤í–‰ë¶€
# -------------------------
def main():
    import tkinter as tk

    # (ê¶Œì¥) ì…ë ¥ì¥ì¹˜ ì§„ë‹¨ ë¡œê·¸(ë¬¸ì œ ìƒê¸¸ ë•Œë§Œ ì¼œë„ ë¨)
    # print(sd.query_devices())
    # print("default:", sd.default.device)

    root = tk.Tk()
    app = STApp(root)

    # ì‹œì‘ ì•ˆë‚´(í‚¤ ì—†ìœ¼ë©´ ì—¬ê¸°ì„œ ë°”ë¡œ íŒì—…)
    try:
        _ = get_openai_key()
        app.log("âœ… OpenAI Key í™•ì¸ ì™„ë£Œ(í™˜ê²½ë³€ìˆ˜)")
    except Exception as e:
        app.log(f"âŒ OpenAI Key ë¬¸ì œ: {e}")
        app.popup_error("OpenAI í‚¤ ì˜¤ë¥˜", str(e))

    root.mainloop()

if __name__ == "__main__":
    main()
```
