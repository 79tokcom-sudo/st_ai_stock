```python
# -*- coding: utf-8 -*-
"""
ì²œì‹ ëŒ€ì™• ST AI ì¦ê¶Œ v18.1 (ì›ì†ŒìŠ¤ 1íŒŒì¼)
âœ… ì´ë²ˆ ë²„ì „ í•µì‹¬(ìš”ì²­ â€œë‹¤ìŒâ€ ë°˜ì˜):
- ë°©ì†¡ ì‹ ì²­(íšŒì›) â†’ ê´€ë¦¬ì ìŠ¹ì¸(admin/tt1234) â†’ ë°©ì†¡ ë¦¬ìŠ¤íŠ¸ ë…¸ì¶œ â†’ ì‹œì²­(ì…ì¥) ê°€ëŠ¥
- ì±„íŒ…ë°©: ì´ëª¨í‹°ì½˜ + ë³„í’ì„  + ì• ë‹ˆë©”ì´ì…˜ + ì†ë„ì œí•œ(2ì´ˆ)
- ì¶”ì²œì¢…ëª©: êµ­ë‚´/ë¯¸êµ­ ë¶„ë¦¬ + ì¢…ëª© ë”ë¸”í´ë¦­ â†’ ì‹œì¥ê°€ ë§¤ìˆ˜ì°½(í˜„ì¬ê°€ ìˆ˜ì •ë¶ˆê°€) + 5/10/50/ì „ì•¡ ë²„íŠ¼
- ë³´ìœ ì¢…ëª©/ì£¼ë¬¸ë‚´ì—­/ìˆ˜ìµë¶„ì„/ìµœê·¼ê²€ìƒ‰ íƒ­
- í”„ë¦¬ì§•(ë©ˆì¶¤) ë°©ì§€: í˜„ì¬ê°€ ì¡°íšŒëŠ” ë¹„ë™ê¸°(ìŠ¤ë ˆë“œ) + íƒ€ì„ì•„ì›ƒ
- ì˜¤ë¥˜ ë°œìƒ ì‹œ Firestore error_logs ì €ì¥ + error_fixes ê·œì¹™ ê¸°ë°˜ ìë™ì™„í™”(ê¸°ë³¸ ë£° íƒ‘ì¬)

ì„¤ì¹˜:
pip install firebase-admin websockets yfinance

ì‹¤í–‰:
python st_ai_stock_v18_1.py

ì£¼ì˜:
- Jupyterì—ì„œ ì‹¤í–‰í•˜ë©´ ë©ˆì¶¤/ì—ëŸ¬ê°€ ë‚  ìˆ˜ ìˆìœ¼ë‹ˆ ë°˜ë“œì‹œ í„°ë¯¸ë„ì—ì„œ ì‹¤í–‰ ê¶Œì¥
"""

import os, sys, json, time, threading, queue, traceback, uuid, random
from datetime import datetime, timedelta, date

import tkinter as tk
from tkinter import ttk, messagebox, simpledialog

import firebase_admin
from firebase_admin import credentials, firestore

import yfinance as yf

import asyncio
import websockets
import re


# =========================
# ì„¤ì •
# =========================
APP_NAME = "ì²œì‹ ëŒ€ì™• ST AI ì¦ê¶Œ"
APP_VERSION = "v18.1"

SERVICE_ACCOUNT_JSON = "staidb-firebase-adminsdk-fbsvc-d3ba815ea4.json"

USD_TO_KRW_RATE = 1300.0

CHAT_HOST = "127.0.0.1"
CHAT_PORT = 8765
CHAT_URL = f"ws://{CHAT_HOST}:{CHAT_PORT}"

# Firestore ì»¬ë ‰ì…˜
COL_MEMBERS = "members"
COL_PAYMENT_REQUESTS = "payment_requests"
COL_CHAT_MESSAGES = "chat_messages"
COL_SYSTEM_CHECKS = "system_checks"
COL_ORDERS = "orders"
COL_DIVIDENDS = "dividends"

# ë°©ì†¡ ìŠ¹ì¸í˜• ì»¬ë ‰ì…˜(ì¶”ê°€)
COL_BROADCAST_REQUESTS = "broadcast_requests"
COL_BROADCASTS = "broadcasts"

# ì˜¤ë¥˜ ìë™í™” ì»¬ë ‰ì…˜
ERROR_LOGS_COLLECTION = "error_logs"
ERROR_FIXES_COLLECTION = "error_fixes"

PLAN_1M_PRICE = 300000
PLAN_1Y_PRICE = 3000000
KB_BANK_ACCOUNT_TEXT = "KBì€í–‰ 061701-04-086503 (ì˜ˆê¸ˆì£¼: ì‹ ì£¼í—Œ)"
CONTACT_TEXT = "ì…ê¸ˆí›„ 010 7608 9847 ì—°ë½ë°”ëë‹ˆë‹¤."
BOTTOM_ACCOUNT_TEXT = (
    f"ìœ ë£Œê²°ì œ/ì¶©ì „: {KB_BANK_ACCOUNT_TEXT} | "
    f"1ë‹¬ {PLAN_1M_PRICE:,}ì› / 1ë…„ {PLAN_1Y_PRICE:,}ì› | {CONTACT_TEXT}"
)

EMOTES = ["ğŸ˜€", "ğŸ˜‚", "ğŸ˜", "ğŸ˜­", "ğŸ˜¡", "ğŸ‘", "ğŸ”¥", "ğŸ‰", "ğŸ’", "ğŸš€", "ğŸ™", "ğŸ’°", "ğŸ“ˆ", "ğŸ“‰"]
LOBBY = "ì „ì²´ë°©"
CHAT_MIN_INTERVAL_SEC = 2.0
BALLOON_PRICE = 100.0

DEFAULT_KR = [
    ("ì‚¼ì„±ì „ì", "005930.KS"), ("SKí•˜ì´ë‹‰ìŠ¤", "000660.KS"), ("í˜„ëŒ€ì°¨", "005380.KS"),
    ("ì¹´ì¹´ì˜¤", "035720.KS"), ("ì…€íŠ¸ë¦¬ì˜¨", "068270.KS"), ("LGí™”í•™", "051910.KS"),
    ("KT", "030200.KS"), ("ë§¤ì»¤ìŠ¤", "093520.KQ"), ("ì—˜ì•¤ì”¨ë°”ì´ì˜¤", "290650.KQ"), ("ì™€ì´ë°”ì´ì˜¤ë¡œì§ìŠ¤", "338840.KQ")
]
DEFAULT_US = [
    ("ì—”ë¹„ë””ì•„", "NVDA"), ("ì• í”Œ", "AAPL"), ("ì½”ìŠ¤íŠ¸ì½”", "COST"), ("ì‹œìŠ¤ì½”", "CSCO"),
    ("í•€í„°ë ˆìŠ¤íŠ¸", "PINS"), ("ì½”ë…¸ì½”í•„ë¦½ìŠ¤", "COP"), ("ë¨¼ë°ì´ë‹·ì»´", "MNDY"),
    ("ì›Œë„ˆë¸Œë¡œìŠ¤", "WBD"), ("VSAT", "VSAT"), ("HAE", "HAE")
]


# =========================
# ìœ í‹¸
# =========================
def now_str():
    return datetime.now().strftime("%Y-%m-%d %H:%M:%S")

def month_key():
    return date.today().strftime("%Y-%m")

def safe_float(x, default=0.0):
    try:
        return float(x)
    except Exception:
        return default

def safe_int(x, default=0):
    try:
        return int(float(x))
    except Exception:
        return default

def init_firestore():
    if not firebase_admin._apps:
        if not os.path.exists(SERVICE_ACCOUNT_JSON):
            raise FileNotFoundError(f"ì„œë¹„ìŠ¤ê³„ì • JSON ì—†ìŒ: {SERVICE_ACCOUNT_JSON}")
        cred = credentials.Certificate(SERVICE_ACCOUNT_JSON)
        firebase_admin.initialize_app(cred)
    return firestore.client()

def fetch_market_price(ticker: str):
    """ì‹œì¥ê°€(ìˆ˜ì • ë¶ˆê°€). ë„¤íŠ¸ì›Œí¬ ì‹¤íŒ¨ ì‹œ None"""
    try:
        t = yf.Ticker(ticker)
        fi = getattr(t, "fast_info", None)
        if fi:
            lp = fi.get("last_price") or fi.get("regular_market_price") or fi.get("regularMarketPrice")
            if lp and float(lp) > 0:
                return float(lp)
    except Exception:
        pass
    try:
        t = yf.Ticker(ticker)
        info = getattr(t, "info", None)
        if info:
            p = info.get("regularMarketPrice") or info.get("currentPrice")
            if p and float(p) > 0:
                return float(p)
    except Exception:
        pass
    return None

def compute_signal_simple(ticker: str):
    """BUY/SELL/HOLD (ë‹¨ìˆœ ê·œì¹™)"""
    try:
        df = yf.download(ticker, period="20d", interval="1d", progress=False)
        if df is None or df.empty or "Close" not in df.columns:
            return "HOLD"
        close = df["Close"].dropna()
        if len(close) < 8:
            return "HOLD"
        d = close.diff().dropna()
        last3 = d.iloc[-3:]
        last7 = d.iloc[-7:]
        if (last3 < 0).all():
            return "SELL"
        if int((last7 > 0).sum()) >= 5:
            return "BUY"
        return "HOLD"
    except Exception:
        return "HOLD"

def fmt_remaining(seconds: int) -> str:
    if seconds <= 0:
        return "ë§Œë£Œ"
    d = seconds // 86400
    h = (seconds % 86400) // 3600
    m = (seconds % 3600) // 60
    s = seconds % 60
    if d > 0:
        return f"{d}ì¼ {h}ì‹œê°„ {m}ë¶„"
    return f"{h:02d}:{m:02d}:{s:02d}"


# =========================
# ì˜¤ë¥˜ ì €ì¥ + ì˜¤í† í”½ìŠ¤(ê¸°ë³¸ ë£°)
# =========================
def log_error_to_db(db, where: str, exc: BaseException, extra: dict | None = None):
    try:
        tb = traceback.format_exc()
        doc = {
            "time": now_str(),
            "where": where,
            "error_type": type(exc).__name__,
            "error_msg": str(exc),
            "traceback": tb,
            "extra": extra or {},
            "created_at": firestore.SERVER_TIMESTAMP,
        }
        db.collection(ERROR_LOGS_COLLECTION).add(doc)
    except Exception:
        pass

def ensure_default_fix_rules(db):
    defaults = [
        {
            "pattern": r"An exception has occurred, use %tb",
            "action": "SHOW_GUIDE",
            "note": "Jupyterì—ì„œ Tkinter ì‹¤í–‰ ì‹œ ë°œìƒ ê°€ëŠ¥. í„°ë¯¸ë„ ì‹¤í–‰ ê¶Œì¥."
        },
        {
            "pattern": r"Glyph .* missing from font",
            "action": "SET_MPL_FONT",
            "note": "matplotlib í•œê¸€ í°íŠ¸ ë¯¸ì„¤ì • ê²½ê³  ì™„í™”"
        },
        {
            "pattern": r"KeyboardInterrupt",
            "action": "IGNORE",
            "note": "ì‚¬ìš©ì ì¤‘ë‹¨ì€ ë¬´ì‹œ"
        },
    ]
    try:
        # ì´ë¯¸ ìˆìœ¼ë©´ ìŠ¤í‚µ
        snap = db.collection(ERROR_FIXES_COLLECTION).limit(1).get()
        if snap:
            return
        for d in defaults:
            db.collection(ERROR_FIXES_COLLECTION).add(d)
    except Exception:
        pass

def load_fix_rules(db):
    rules = []
    try:
        for s in db.collection(ERROR_FIXES_COLLECTION).stream():
            d = s.to_dict() or {}
            if d.get("pattern") and d.get("action"):
                rules.append(d)
    except Exception:
        pass
    return rules

def apply_auto_fix(app, error_text: str):
    rules = getattr(app, "_fix_rules", None)
    if rules is None:
        app._fix_rules = load_fix_rules(app.db)
        rules = app._fix_rules

    for r in rules:
        try:
            if re.search(r["pattern"], error_text, re.IGNORECASE):
                action = r.get("action")
                if action == "SHOW_GUIDE":
                    messagebox.showwarning(
                        "ì‹¤í–‰ ì•ˆë‚´",
                        "í˜„ì¬ í™˜ê²½ì´ Jupyterì¼ ê°€ëŠ¥ì„±ì´ í½ë‹ˆë‹¤.\n"
                        "í„°ë¯¸ë„ì—ì„œ python íŒŒì¼ë¡œ ì‹¤í–‰í•˜ì„¸ìš”.\n"
                        "ì˜ˆ: python st_ai_stock_v18_1.py"
                    )
                    return True
                if action == "SET_MPL_FONT":
                    try:
                        import matplotlib
                        matplotlib.rcParams["font.family"] = "Malgun Gothic"
                        matplotlib.rcParams["axes.unicode_minus"] = False
                    except Exception:
                        pass
                    return True
                if action == "IGNORE":
                    return True
        except Exception:
            continue
    return False


# =========================
# Firestore Repo
# =========================
class RepoFS:
    def __init__(self, db):
        self.db = db

    def ensure_admin_bootstrap(self):
        ref = self.db.collection(COL_MEMBERS).document("admin")
        snap = ref.get()
        if snap.exists:
            d = snap.to_dict() or {}
            patch = {}
            if d.get("role") != "admin":
                patch["role"] = "admin"
                patch["grade"] = "ADMIN"
            if not d.get("subscription_expiry"):
                patch["subscription_expiry"] = "2099-12-31 23:59:59"
            if patch:
                patch["updated_at"] = firestore.SERVER_TIMESTAMP
                ref.set(patch, merge=True)
            return

        ref.set({
            "user_id": "admin",
            "password": "tt1234",
            "phone": "",
            "role": "admin",
            "grade": "ADMIN",
            "payment_status": "ACTIVE",
            "subscription_expiry": "2099-12-31 23:59:59",

            "balance_krw": 0.0,
            "balance_usd": 0.0,
            "balance_cash": 0.0,

            "holdings": {},
            "recent_stocks": [],

            "created_at": firestore.SERVER_TIMESTAMP,
            "updated_at": firestore.SERVER_TIMESTAMP
        }, merge=True)

    def member_get(self, user_id: str):
        snap = self.db.collection(COL_MEMBERS).document(user_id).get()
        return snap.to_dict() if snap.exists else None

    def member_create(self, user_id: str, pw: str, phone: str = ""):
        ref = self.db.collection(COL_MEMBERS).document(user_id)
        if ref.get().exists:
            return False, "ì´ë¯¸ ì¡´ì¬í•˜ëŠ” ID"
        start = datetime.now()
        expiry = start + timedelta(hours=24)
        ref.set({
            "user_id": user_id,
            "password": pw,
            "phone": phone,
            "role": "user",
            "grade": "BASIC",
            "payment_status": "TRIAL",
            "trial_start": start.strftime("%Y-%m-%d %H:%M:%S"),
            "subscription_expiry": expiry.strftime("%Y-%m-%d %H:%M:%S"),

            "balance_krw": 0.0,
            "balance_usd": 0.0,
            "balance_cash": 0.0,

            "holdings": {},
            "recent_stocks": [],

            "created_at": firestore.SERVER_TIMESTAMP,
            "updated_at": firestore.SERVER_TIMESTAMP
        }, merge=True)
        return True, "íšŒì›ê°€ì… ì™„ë£Œ(ë¬´ë£Œ 24ì‹œê°„ ì‹œì‘)"

    def verify_login(self, user_id: str, pw: str):
        m = self.member_get(user_id)
        if not m:
            return False
        return (m.get("password") == pw)

    def member_update(self, user_id: str, patch: dict):
        self.db.collection(COL_MEMBERS).document(user_id).set({**patch, "updated_at": firestore.SERVER_TIMESTAMP}, merge=True)

    def create_payment_request(self, user_id: str, typ: str, depositor: str, amount: float, phone: str, memo: str):
        self.db.collection(COL_PAYMENT_REQUESTS).add({
            "user_id": user_id,
            "type": typ,  # PLAN_1M/PLAN_1Y/KRW/USD/CASH
            "depositor": depositor,
            "amount": float(amount),
            "phone": phone,
            "memo": memo,
            "status": "PENDING",
            "bank": KB_BANK_ACCOUNT_TEXT,
            "contact": CONTACT_TEXT,
            "client_time": now_str(),
            "created_at": firestore.SERVER_TIMESTAMP
        })

    def add_order(self, order: dict):
        self.db.collection(COL_ORDERS).add(order)

    def list_orders_for_user(self, user_id: str, limit=200):
        out = []
        q = (self.db.collection(COL_ORDERS)
             .where("user_id", "==", user_id)
             .order_by("created_at", direction=firestore.Query.DESCENDING)
             .limit(limit))
        for s in q.stream():
            d = s.to_dict() or {}
            d["_id"] = s.id
            out.append(d)
        return out

    def sum_dividends_month(self, user_id: str, month: str):
        total = 0.0
        try:
            q = (self.db.collection(COL_DIVIDENDS)
                 .where("user_id", "==", user_id)
                 .where("month", "==", month)
                 .limit(200))
            for s in q.stream():
                d = s.to_dict() or {}
                total += float(d.get("amount", 0.0))
        except Exception:
            pass
        return total

    def save_chat_message(self, room: str, msg_type: str, user: str, message: str = "", gift: str = "", count: int = 0):
        try:
            self.db.collection(COL_CHAT_MESSAGES).add({
                "room": room,
                "msg_type": msg_type,
                "user": user,
                "message": message,
                "gift": gift,
                "count": int(count),
                "time": now_str(),
                "created_at": firestore.SERVER_TIMESTAMP
            })
        except Exception:
            pass

    # ===== ë°©ì†¡ ì‹ ì²­/ìŠ¹ì¸í˜•(DB ê¸°ë°˜) =====
    def create_broadcast_request(self, user_id: str, title: str, category: str = "ì£¼ì‹", memo: str = ""):
        self.db.collection(COL_BROADCAST_REQUESTS).add({
            "user_id": user_id,
            "title": title,
            "category": category,
            "memo": memo,
            "status": "PENDING",  # PENDING / APPROVED / REJECTED
            "created_time": now_str(),
            "created_at": firestore.SERVER_TIMESTAMP
        })

    def list_broadcast_requests(self, status="PENDING", limit=200):
        out = []
        q = (self.db.collection(COL_BROADCAST_REQUESTS)
             .where("status", "==", status)
             .order_by("created_at", direction=firestore.Query.DESCENDING)
             .limit(limit))
        for s in q.stream():
            d = s.to_dict() or {}
            d["_id"] = s.id
            out.append(d)
        return out

    def approve_broadcast_request(self, request_id: str, admin_id: str):
        ref = self.db.collection(COL_BROADCAST_REQUESTS).document(request_id)
        snap = ref.get()
        if not snap.exists:
            return False, "ì‹ ì²­ì„œ ì—†ìŒ"
        req = snap.to_dict() or {}
        if req.get("status") != "PENDING":
            return False, "ì´ë¯¸ ì²˜ë¦¬ëœ ì‹ ì²­ì„œ"

        bid = "db" + uuid.uuid4().hex[:10]
        self.db.collection(COL_BROADCASTS).document(bid).set({
            "broadcast_id": bid,
            "title": req.get("title", "ë¬´ì œ"),
            "category": req.get("category", "ì£¼ì‹"),
            "host": req.get("user_id", ""),
            "status": "ACTIVE",  # ACTIVE / ENDED
            "approved_by": admin_id,
            "approved_time": now_str(),
            "created_time": req.get("created_time", now_str()),
            "created_at": firestore.SERVER_TIMESTAMP
        }, merge=True)

        ref.set({
            "status": "APPROVED",
            "approved_by": admin_id,
            "approved_time": now_str(),
            "approved_broadcast_id": bid,
            "updated_at": firestore.SERVER_TIMESTAMP
        }, merge=True)
        return True, f"ìŠ¹ì¸ ì™„ë£Œ(ë°©ì†¡ID: {bid})"

    def list_active_broadcasts(self, limit=200):
        out = []
        q = (self.db.collection(COL_BROADCASTS)
             .where("status", "==", "ACTIVE")
             .order_by("created_at", direction=firestore.Query.DESCENDING)
             .limit(limit))
        for s in q.stream():
            d = s.to_dict() or {}
            d["_id"] = s.id
            out.append(d)
        return out

    def end_broadcast(self, broadcast_id: str, user_id: str):
        ref = self.db.collection(COL_BROADCASTS).document(broadcast_id)
        snap = ref.get()
        if not snap.exists:
            return False, "ë°©ì†¡ ì—†ìŒ"
        d = snap.to_dict() or {}
        if d.get("host") != user_id:
            return False, "í˜¸ìŠ¤íŠ¸ë§Œ ì¢…ë£Œ ê°€ëŠ¥"
        ref.set({
            "status": "ENDED",
            "ended_time": now_str(),
            "updated_at": firestore.SERVER_TIMESTAMP
        }, merge=True)
        return True, "ë°©ì†¡ ì¢…ë£Œ ì™„ë£Œ"


# =========================
# Embedded WebSocket Server (ì±„íŒ…/ì„ ë¬¼/ë°©ì†¡ë°©)
# =========================
class EmbeddedServer:
    def __init__(self, repo: RepoFS, host=CHAT_HOST, port=CHAT_PORT):
        self.repo = repo
        self.host = host
        self.port = port
        self.thread = None

        self.lock = threading.Lock()
        self.rooms = {}           # room -> set(ws)
        self.ws_room = {}         # ws -> room
        self.ws_last_send = {}    # ws -> ts

    def start(self):
        if self.thread and self.thread.is_alive():
            return

        def runner():
            try:
                asyncio.run(self._main())
            except OSError:
                # ì´ë¯¸ ì‹¤í–‰ì¤‘
                pass
            except Exception:
                pass

        self.thread = threading.Thread(target=runner, daemon=True)
        self.thread.start()

    def _can_send(self, ws):
        now = time.time()
        last = self.ws_last_send.get(ws, 0.0)
        if (now - last) < CHAT_MIN_INTERVAL_SEC:
            return False
        self.ws_last_send[ws] = now
        return True

    async def _send(self, ws, payload: dict):
        try:
            await ws.send(json.dumps(payload, ensure_ascii=False))
        except Exception:
            pass

    async def _broadcast_room(self, room: str, payload: dict):
        with self.lock:
            clients = list(self.rooms.get(room, set()))
        msg = json.dumps(payload, ensure_ascii=False)
        for c in clients:
            try:
                await c.send(msg)
            except Exception:
                pass

    async def _handler(self, ws):
        room = LOBBY
        with self.lock:
            self.rooms.setdefault(room, set()).add(ws)
            self.ws_room[ws] = room

        try:
            async for raw in ws:
                try:
                    payload = json.loads(raw)
                except Exception:
                    payload = {"type": "chat", "room": room, "user": "unknown", "message": raw}

                typ = payload.get("type")

                if typ == "join":
                    new_room = (payload.get("room") or LOBBY).strip() or LOBBY
                    if new_room != room:
                        with self.lock:
                            self.rooms.get(room, set()).discard(ws)
                            self.rooms.setdefault(new_room, set()).add(ws)
                            self.ws_room[ws] = new_room
                        room = new_room
                    continue

                if typ in ("chat", "gift"):
                    if not self._can_send(ws):
                        await self._send(ws, {"type": "system", "room": room, "time": now_str(), "message": "ì „ì†¡ ì†ë„ ì œí•œ(2ì´ˆ)"})
                        continue

                    room2 = (payload.get("room") or room).strip() or room
                    user = str(payload.get("user", "unknown"))[:40]
                    payload["room"] = room2
                    payload["time"] = now_str()

                    if typ == "chat":
                        self.repo.save_chat_message(room2, "chat", user=user, message=str(payload.get("message", ""))[:500])
                    else:
                        cnt = int(payload.get("count", 1))
                        gift = str(payload.get("gift", "ğŸ"))[:20]
                        self.repo.save_chat_message(room2, "gift", user=user, gift=gift, count=cnt)

                    await self._broadcast_room(room2, payload)
                    continue

        except Exception:
            pass
        finally:
            with self.lock:
                cr = self.ws_room.get(ws, LOBBY)
                self.rooms.get(cr, set()).discard(ws)
                self.ws_room.pop(ws, None)
                self.ws_last_send.pop(ws, None)

    async def _main(self):
        async with websockets.serve(self._handler, self.host, self.port, ping_interval=20, ping_timeout=20):
            while True:
                await asyncio.sleep(3600)


# =========================
# ì•±
# =========================
class STApp:
    def __init__(self, root):
        self.root = root
        self.root.title(f"{APP_NAME} {APP_VERSION}")
        self.root.geometry("1900x1020")
        self.root.configure(bg="white")

        self.db = init_firestore()
        ensure_default_fix_rules(self.db)  # ì˜¤í† í”½ìŠ¤ ê¸°ë³¸ë£°
        self._fix_rules = load_fix_rules(self.db)

        self.repo = RepoFS(self.db)
        self.repo.ensure_admin_bootstrap()

        self.server = EmbeddedServer(self.repo)
        self.server.start()

        # ìƒíƒœ
        self.current_user = "guest"
        self.role = "guest"
        self.grade = "GUEST"

        self.user_data = {
            "balance_krw": 0.0,
            "balance_usd": 0.0,
            "balance_cash": 0.0,
            "holdings": {},
            "recent_stocks": [],
            "subscription_expiry": ""
        }
        self.orders_cache = []

        # worker
        self.work_q = queue.Queue()
        self.ui_q = queue.Queue()
        self.worker = threading.Thread(target=self._worker_loop, daemon=True)
        self.worker.start()

        # ws client
        self.ws_connected = False
        self.ws_thread = None
        self.ws_send_q = queue.Queue()
        self.ws_ui_q = queue.Queue()
        self.chat_room = LOBBY
        self.last_send_ts = 0.0

        # ë¼ì´ë¸Œ DB ë¦¬ìŠ¤íŠ¸ ìºì‹œ
        self.live_db_items = []

        self._build_login()
        self.root.after(150, self._tick)

    # ---------------- worker ----------------
    def _worker_loop(self):
        while True:
            job = self.work_q.get()
            if job is None:
                break
            fn, args = job
            try:
                res = fn(*args)
                self.ui_q.put(("OK", fn.__name__, res))
            except Exception as e:
                self.ui_q.put(("ERR", fn.__name__, e))

    def _tick(self):
        try:
            while not self.ui_q.empty():
                tag, name, res = self.ui_q.get_nowait()
                if tag == "OK":
                    if name == "job_refresh_signals":
                        self._apply_signals(res)
                    elif name == "job_refresh_prices":
                        self._apply_prices(res)
                else:
                    # worker error
                    log_error_to_db(self.db, f"worker:{name}", res, {})
                    apply_auto_fix(self, str(res))
        except Exception:
            pass

        try:
            while not self.ws_ui_q.empty():
                payload = self.ws_ui_q.get_nowait()
                self._on_ws_payload(payload)
        except Exception:
            pass

        self.root.after(150, self._tick)

    # ---------------- ws ----------------
    def connect_ws(self):
        if self.ws_connected:
            return

        def loop():
            async def run():
                try:
                    async with websockets.connect(CHAT_URL, ping_interval=20, ping_timeout=20) as ws:
                        self.ws_connected = True

                        async def sender():
                            while True:
                                msg = await asyncio.get_event_loop().run_in_executor(None, self.ws_send_q.get)
                                await ws.send(msg)

                        async def receiver():
                            while True:
                                data = await ws.recv()
                                try:
                                    payload = json.loads(data)
                                except Exception:
                                    payload = {"type": "system_local", "message": str(data)}
                                self.ws_ui_q.put(payload)

                        await ws.send(json.dumps({"type": "join", "room": self.chat_room}, ensure_ascii=False))
                        await asyncio.gather(sender(), receiver())
                except Exception as e:
                    self.ws_connected = False
                    self.ws_ui_q.put({"type": "system_local", "message": f"WS ëŠê¹€: {e}"})

            asyncio.run(run())

        self.ws_thread = threading.Thread(target=loop, daemon=True)
        self.ws_thread.start()

    def ws_send(self, payload: dict):
        if not self.ws_connected:
            self.connect_ws()
            time.sleep(0.08)
        self.ws_send_q.put(json.dumps(payload, ensure_ascii=False))

    def _can_send_client(self):
        now = time.time()
        if (now - self.last_send_ts) < CHAT_MIN_INTERVAL_SEC:
            return False
        self.last_send_ts = now
        return True

    # ---------------- UI util ----------------
    def _clear(self):
        for w in self.root.winfo_children():
            w.destroy()

    def total_krw(self):
        return float(self.user_data.get("balance_krw", 0.0)) + float(self.user_data.get("balance_usd", 0.0)) * USD_TO_KRW_RATE

    def refresh_topbar(self):
        if hasattr(self, "top_total_var"):
            self.top_total_var.set(
                f"ì´ê¸ˆì•¡ {self.total_krw():,.0f}ì› | "
                f"KRW {self.user_data['balance_krw']:,.0f} | "
                f"USD {self.user_data['balance_usd']:,.2f} | "
                f"CASH {self.user_data['balance_cash']:,.0f}"
            )
        if hasattr(self, "kpi_orders_var"):
            self._refresh_kpis()

    def _refresh_kpis(self):
        m = month_key()
        month_orders = [o for o in (self.orders_cache or []) if str(o.get("time", "")).startswith(m)]
        order_count = len(month_orders)
        realized = sum(float(o.get("realized_pnl", 0.0)) for o in month_orders if o.get("side") == "SELL")
        div_total = 0.0
        if self.current_user != "guest":
            div_total = float(self.repo.sum_dividends_month(self.current_user, m))

        self.kpi_orders_var.set(f"ì´ë²ˆë‹¬ ì£¼ë¬¸: {order_count}ê±´")
        self.kpi_pnl_var.set(f"ì´ë²ˆë‹¬ íŒë§¤ìˆ˜ìµ: {realized:,.0f}ì›")
        self.kpi_div_var.set(f"ì´ë²ˆë‹¬ ë°°ë‹¹ê¸ˆ: {div_total:,.0f}ì›")

    def _push_recent(self, name: str, ticker: str):
        recent = list(self.user_data.get("recent_stocks", []))
        recent = [x for x in recent if x.get("ticker") != ticker]
        recent.insert(0, {"name": name, "ticker": ticker, "time": now_str()})
        recent = recent[:20]
        self.user_data["recent_stocks"] = recent
        self._render_recent()
        if self.current_user != "guest":
            self.repo.member_update(self.current_user, {"recent_stocks": recent})

    # ---------------- í”„ë¦¬ì§• ë°©ì§€: ê°€ê²© ë¹„ë™ê¸° ----------------
    def fetch_price_async(self, ticker: str, callback, timeout_sec: float = 4.0):
        done = {"ok": False}

        def run():
            try:
                p = fetch_market_price(ticker)
            except Exception as e:
                log_error_to_db(self.db, "fetch_market_price", e, {"ticker": ticker})
                apply_auto_fix(self, str(e))
                p = None
            if not done["ok"]:
                done["ok"] = True
                callback(p)

        t = threading.Thread(target=run, daemon=True)
        t.start()

        def timeout_fire():
            if not done["ok"]:
                done["ok"] = True
                callback(None)
        self.root.after(int(timeout_sec * 1000), timeout_fire)

    # ---------------- Login UI ----------------
    def _build_login(self):
        self._clear()
        box = tk.Frame(self.root, bg="white")
        box.pack(expand=True)

        tk.Label(box, text=APP_NAME, bg="white", fg="#0064FF",
                 font=("ë§‘ì€ ê³ ë”•", 44, "bold")).pack(pady=16)

        frm = tk.Frame(box, bg="white")
        frm.pack(pady=8)

        tk.Label(frm, text="ID", bg="white", fg="#0064FF", font=("ë§‘ì€ ê³ ë”•", 14, "bold")).grid(row=0, column=0, padx=10, pady=10, sticky="e")
        self.id_entry = tk.Entry(frm, font=("ë§‘ì€ ê³ ë”•", 14), width=28)
        self.id_entry.grid(row=0, column=1, padx=10, pady=10)

        tk.Label(frm, text="PW", bg="white", fg="#0064FF", font=("ë§‘ì€ ê³ ë”•", 14, "bold")).grid(row=1, column=0, padx=10, pady=10, sticky="e")
        self.pw_entry = tk.Entry(frm, font=("ë§‘ì€ ê³ ë”•", 14), width=28, show="*")
        self.pw_entry.grid(row=1, column=1, padx=10, pady=10)

        btns = tk.Frame(box, bg="white")
        btns.pack(pady=10)

        tk.Button(btns, text="ë¡œê·¸ì¸", bg="#0064FF", fg="white",
                  font=("ë§‘ì€ ê³ ë”•", 16, "bold"), width=10, command=self.login).grid(row=0, column=0, padx=6)
        tk.Button(btns, text="íšŒì›ê°€ì…", bg="#111111", fg="white",
                  font=("ë§‘ì€ ê³ ë”•", 16, "bold"), width=10, command=self.signup).grid(row=0, column=1, padx=6)
        tk.Button(btns, text="ê²ŒìŠ¤íŠ¸", bg="white", fg="#0064FF",
                  font=("ë§‘ì€ ê³ ë”•", 16, "bold"), width=10, command=self.guest).grid(row=0, column=2, padx=6)

        tk.Label(box, text=BOTTOM_ACCOUNT_TEXT, bg="white", fg="#666666", font=("ë§‘ì€ ê³ ë”•", 12)).pack(pady=12)

    def guest(self):
        self.current_user = "guest"
        self.role = "guest"
        self.grade = "GUEST"
        self.user_data = {
            "balance_krw": 0.0,
            "balance_usd": 0.0,
            "balance_cash": 0.0,
            "holdings": {},
            "recent_stocks": [],
            "subscription_expiry": ""
        }
        self.orders_cache = []
        self._build_main()

    def signup(self):
        uid = self.id_entry.get().strip()
        pw = self.pw_entry.get().strip()
        if not uid or not pw:
            messagebox.showwarning("ì˜¤ë¥˜", "ID/PW ì…ë ¥")
            return
        phone = simpledialog.askstring("íšŒì›ê°€ì…", "íœ´ëŒ€í°(ì„ íƒ):") or ""
        ok, msg = self.repo.member_create(uid, pw, phone=phone)
        messagebox.showinfo("ê²°ê³¼", msg)

    def login(self):
        uid = self.id_entry.get().strip()
        pw = self.pw_entry.get().strip()
        if not uid or not pw:
            messagebox.showwarning("ì˜¤ë¥˜", "ID/PW ì…ë ¥")
            return
        if not self.repo.verify_login(uid, pw):
            messagebox.showerror("ì‹¤íŒ¨", "ë¡œê·¸ì¸ ì‹¤íŒ¨")
            return

        m = self.repo.member_get(uid) or {}
        self.current_user = uid
        self.role = m.get("role", "user")
        self.grade = m.get("grade", "BASIC")

        self.user_data = {
            "balance_krw": float(m.get("balance_krw", 0.0)),
            "balance_usd": float(m.get("balance_usd", 0.0)),
            "balance_cash": float(m.get("balance_cash", 0.0)),
            "holdings": dict(m.get("holdings", {}) or {}),
            "recent_stocks": list(m.get("recent_stocks", []) or []),
            "subscription_expiry": str(m.get("subscription_expiry") or "")
        }
        self.orders_cache = self.repo.list_orders_for_user(uid, limit=200)
        self._build_main()

    # ---------------- Main UI ----------------
    def _build_main(self):
        self._clear()
        self.connect_ws()

        header = tk.Frame(self.root, bg="#0064FF")
        header.pack(fill="x")

        tk.Label(header, text=f"{APP_NAME} | {self.current_user} ({self.role}/{self.grade})",
                 bg="#0064FF", fg="white", font=("ë§‘ì€ ê³ ë”•", 15, "bold")).pack(side="left", padx=14, pady=10)

        self.top_total_var = tk.StringVar(value="")  # refresh_topbar()ì—ì„œ ì±„ì›€
        tk.Label(header, textvariable=self.top_total_var, bg="#0064FF", fg="white",
                 font=("ë§‘ì€ ê³ ë”•", 12, "bold")).pack(side="left", padx=18)

        right = tk.Frame(header, bg="#0064FF")
        right.pack(side="right", padx=10)

        tk.Button(right, text="ì¶©ì „í•˜ê¸°", bg="white", fg="#0064FF", bd=0,
                  font=("ë§‘ì€ ê³ ë”•", 12, "bold"), command=self.open_charge_popup).pack(side="right", padx=8)
        tk.Button(right, text="ê°€ê²©ê°±ì‹ ", bg="white", fg="#0064FF", bd=0,
                  font=("ë§‘ì€ ê³ ë”•", 12, "bold"), command=self.refresh_prices).pack(side="right", padx=8)

        body = tk.Frame(self.root, bg="white")
        body.pack(fill="both", expand=True, padx=12, pady=12)

        # KPI
        kpi = tk.Frame(body, bg="white")
        kpi.pack(fill="x", pady=(0, 10))

        self.kpi_orders_var = tk.StringVar()
        self.kpi_pnl_var = tk.StringVar()
        self.kpi_div_var = tk.StringVar()
        self._refresh_kpis()

        def card(textvar):
            f = tk.Frame(kpi, bg="#f3f5f7")
            f.pack(side="left", padx=8)
            tk.Label(f, textvariable=textvar, bg="#f3f5f7", fg="#111111",
                     font=("ë§‘ì€ ê³ ë”•", 14, "bold"), width=22, anchor="w").pack(padx=10, pady=10)

        card(self.kpi_orders_var)
        card(self.kpi_pnl_var)
        card(self.kpi_div_var)

        # Notebook
        self.nb = ttk.Notebook(body)
        self.nb.pack(fill="both", expand=True)

        self.tab_kr = tk.Frame(self.nb, bg="white")
        self.tab_us = tk.Frame(self.nb, bg="white")
        self.tab_analysis = tk.Frame(self.nb, bg="white")
        self.tab_recent = tk.Frame(self.nb, bg="white")
        self.tab_port = tk.Frame(self.nb, bg="white")
        self.tab_orders = tk.Frame(self.nb, bg="white")
        self.tab_chat = tk.Frame(self.nb, bg="white")
        self.tab_live = tk.Frame(self.nb, bg="white")

        self.nb.add(self.tab_kr, text="êµ­ë‚´ ì¶”ì²œ")
        self.nb.add(self.tab_us, text="ë¯¸êµ­ ì¶”ì²œ")
        self.nb.add(self.tab_analysis, text="ìˆ˜ìµë¶„ì„")
        self.nb.add(self.tab_recent, text="ìµœê·¼ê²€ìƒ‰")
        self.nb.add(self.tab_port, text="ë³´ìœ ì¢…ëª©")
        self.nb.add(self.tab_orders, text="ì£¼ë¬¸ë‚´ì—­")
        self.nb.add(self.tab_chat, text="ì±„íŒ…")
        self.nb.add(self.tab_live, text="ë¼ì´ë¸Œ(ì‹ ì²­/ìŠ¹ì¸)")

        self._build_reco_tab(self.tab_kr, "KR")
        self._build_reco_tab(self.tab_us, "US")
        self._build_analysis_tab(self.tab_analysis)
        self._build_recent_tab(self.tab_recent)
        self._build_portfolio_tab(self.tab_port)
        self._build_orders_tab(self.tab_orders)
        self._build_chat_tab(self.tab_chat)
        self._build_live_tab(self.tab_live)

        tk.Label(body, text=BOTTOM_ACCOUNT_TEXT, bg="white", fg="#666666",
                 font=("ë§‘ì€ ê³ ë”•", 11)).pack(fill="x", pady=(10, 0))

        # ì´ˆê¸° ê°±ì‹ 
        self.refresh_topbar()
        self.work_q.put((self.job_refresh_signals, ()))
        self._render_recent()
        self._render_portfolio({})
        self._render_orders()
        self._set_analysis("ìˆ˜ìµë¶„ì„: ë³´ìœ  ì¢…ëª©ì´ ìˆìœ¼ë©´ 'ê°€ê²©ê°±ì‹ 'ì„ ëˆŒëŸ¬ í˜„ì¬ê°€ ë°˜ì˜ í›„ ìë™ ê³„ì‚°ë©ë‹ˆë‹¤.\n")

    # ---------------- Tabs ----------------
    def _build_reco_tab(self, parent, market: str):
        cols = ("ì¢…ëª©", "í‹°ì»¤", "ì‹ í˜¸")
        tree = ttk.Treeview(parent, columns=cols, show="headings", height=18)
        for c in cols:
            tree.heading(c, text=c)
            tree.column(c, anchor="center")
        tree.column("ì¢…ëª©", width=280)
        tree.column("í‹°ì»¤", width=160)
        tree.column("ì‹ í˜¸", width=120)
        tree.pack(fill="both", expand=True, padx=10, pady=10)

        tree.tag_configure("BUY", foreground="#0a8a2a")
        tree.tag_configure("SELL", foreground="#d02626")
        tree.tag_configure("HOLD", foreground="#666666")

        stocks = DEFAULT_KR if market == "KR" else DEFAULT_US
        for name, tkr in stocks:
            tree.insert("", "end", values=(name, tkr, "HOLD"), tags=("HOLD",))

        def on_dbl(_e):
            sel = tree.selection()
            if not sel:
                return
            name, ticker, sig = tree.item(sel[0], "values")
            self._push_recent(name, ticker)
            self.open_market_buy_popup(name, ticker, sig)

        tree.bind("<Double-Button-1>", on_dbl)

        if market == "KR":
            self.tree_kr = tree
        else:
            self.tree_us = tree

    def _build_analysis_tab(self, parent):
        self.analysis_text = tk.Text(parent, font=("ë§‘ì€ ê³ ë”•", 12))
        self.analysis_text.pack(fill="both", expand=True, padx=10, pady=10)

    def _build_recent_tab(self, parent):
        self.recent_list = tk.Listbox(parent, height=20, font=("ë§‘ì€ ê³ ë”•", 12))
        self.recent_list.pack(fill="both", expand=True, padx=10, pady=10)

    def _build_portfolio_tab(self, parent):
        cols = ("ì¢…ëª©", "í‹°ì»¤", "ìˆ˜ëŸ‰", "í‰ë‹¨", "í˜„ì¬ê°€", "ì†ìµ", "ìˆ˜ìµë¥ (%)")
        self.tree_port = ttk.Treeview(parent, columns=cols, show="headings", height=18)
        for c in cols:
            self.tree_port.heading(c, text=c)
            self.tree_port.column(c, anchor="center")
        self.tree_port.column("ì¢…ëª©", width=260)
        self.tree_port.column("í‹°ì»¤", width=160)
        self.tree_port.column("ìˆ˜ëŸ‰", width=90)
        self.tree_port.column("í‰ë‹¨", width=120)
        self.tree_port.column("í˜„ì¬ê°€", width=120)
        self.tree_port.column("ì†ìµ", width=150)
        self.tree_port.column("ìˆ˜ìµë¥ (%)", width=120)
        self.tree_port.pack(fill="both", expand=True, padx=10, pady=10)

        tk.Button(parent, text="ì„ íƒ ì¢…ëª© ë§¤ë„", bg="#ff3b30", fg="white",
                  font=("ë§‘ì€ ê³ ë”•", 12, "bold"), command=self.sell_selected).pack(pady=6)

    def _build_orders_tab(self, parent):
        cols = ("ì‹œê°„", "êµ¬ë¶„", "ì¢…ëª©", "í‹°ì»¤", "ìˆ˜ëŸ‰", "ì²´ê²°ê°€", "í†µí™”", "ê¸ˆì•¡")
        self.tree_orders = ttk.Treeview(parent, columns=cols, show="headings", height=18)
        for c in cols:
            self.tree_orders.heading(c, text=c)
            self.tree_orders.column(c, anchor="center")
        self.tree_orders.column("ì‹œê°„", width=170)
        self.tree_orders.column("êµ¬ë¶„", width=80)
        self.tree_orders.column("ì¢…ëª©", width=240)
        self.tree_orders.column("í‹°ì»¤", width=150)
        self.tree_orders.column("ìˆ˜ëŸ‰", width=90)
        self.tree_orders.column("ì²´ê²°ê°€", width=120)
        self.tree_orders.column("í†µí™”", width=70)
        self.tree_orders.column("ê¸ˆì•¡", width=120)
        self.tree_orders.pack(fill="both", expand=True, padx=10, pady=10)

        tk.Button(parent, text="ì£¼ë¬¸ë‚´ì—­ ìƒˆë¡œê³ ì¹¨", bg="#0064FF", fg="white",
                  font=("ë§‘ì€ ê³ ë”•", 11, "bold"), command=self.reload_orders).pack(pady=6)

    def _build_chat_tab(self, parent):
        self.chat_text = tk.Text(parent, font=("ë§‘ì€ ê³ ë”•", 11))
        self.chat_text.pack(fill="both", expand=True, padx=10, pady=(10, 6))
        self.chat_text.configure(state="disabled")

        self.chat_canvas = tk.Canvas(parent, bg="", highlightthickness=0, height=260)
        self.chat_canvas.pack(fill="x", padx=10, pady=(0, 6))

        em = tk.Frame(parent, bg="white")
        em.pack(fill="x", padx=10, pady=6)
        for e in EMOTES:
            tk.Button(em, text=e, bg="#f3f5f7",
                      command=lambda x=e: self.chat_entry.insert(tk.END, x)).pack(side="left", padx=3)

        bottom = tk.Frame(parent, bg="white")
        bottom.pack(fill="x", padx=10, pady=(0, 10))
        self.chat_entry = tk.Entry(bottom, font=("ë§‘ì€ ê³ ë”•", 12))
        self.chat_entry.pack(side="left", fill="x", expand=True, padx=(0, 8))
        tk.Button(bottom, text="ì „ì†¡", bg="#0064FF", fg="white", font=("ë§‘ì€ ê³ ë”•", 12, "bold"),
                  command=self.chat_send).pack(side="right", padx=6)
        tk.Button(bottom, text="ğŸˆë³„í’ì„ ", bg="#ff3b30", fg="white", font=("ë§‘ì€ ê³ ë”•", 12, "bold"),
                  command=self.chat_send_balloon).pack(side="right", padx=6)

    def _build_live_tab(self, parent):
        top = tk.Frame(parent, bg="white")
        top.pack(fill="x", padx=10, pady=10)

        tk.Button(top, text="ë°©ì†¡ ì‹ ì²­", bg="#111111", fg="white",
                  font=("ë§‘ì€ ê³ ë”•", 11, "bold"),
                  command=self.live_request).pack(side="left", padx=6)

        tk.Button(top, text="ë°©ì†¡ ë¦¬ìŠ¤íŠ¸ ìƒˆë¡œê³ ì¹¨", bg="#0064FF", fg="white",
                  font=("ë§‘ì€ ê³ ë”•", 11, "bold"),
                  command=self.live_refresh_from_db).pack(side="left", padx=6)

        if self.role == "admin":
            tk.Button(top, text="ê´€ë¦¬ì: ë°©ì†¡ì‹ ì²­ ìŠ¹ì¸", bg="#ff3b30", fg="white",
                      font=("ë§‘ì€ ê³ ë”•", 11, "bold"),
                      command=self.open_broadcast_admin_panel).pack(side="left", padx=6)

        cols = ("ì œëª©", "ì¹´í…Œê³ ë¦¬", "í˜¸ìŠ¤íŠ¸", "ë°©ì†¡ID", "ìƒíƒœ", "ì‹œì‘")
        self.live_tree_db = ttk.Treeview(parent, columns=cols, show="headings", height=18)
        for c in cols:
            self.live_tree_db.heading(c, text=c)
            self.live_tree_db.column(c, anchor="center")
        self.live_tree_db.column("ì œëª©", width=360)
        self.live_tree_db.column("ì¹´í…Œê³ ë¦¬", width=120)
        self.live_tree_db.column("í˜¸ìŠ¤íŠ¸", width=140)
        self.live_tree_db.column("ë°©ì†¡ID", width=160)
        self.live_tree_db.column("ìƒíƒœ", width=90)
        self.live_tree_db.column("ì‹œì‘", width=160)
        self.live_tree_db.pack(fill="both", expand=True, padx=10, pady=(0, 10))

        bar = tk.Frame(parent, bg="white")
        bar.pack(fill="x", padx=10, pady=(0, 10))
        tk.Button(bar, text="ì‹œì²­(ì…ì¥)", bg="#0064FF", fg="white",
                  font=("ë§‘ì€ ê³ ë”•", 11, "bold"),
                  command=self.live_join_selected_db).pack(side="right", padx=6)
        tk.Button(bar, text="ë°©ì†¡ ì¢…ë£Œ(í˜¸ìŠ¤íŠ¸)", bg="#ff3b30", fg="white",
                  font=("ë§‘ì€ ê³ ë”•", 11, "bold"),
                  command=self.live_end_selected_db).pack(side="right", padx=6)

        self.live_refresh_from_db()

    # ---------------- ë°©ì†¡ ì‹ ì²­/ìŠ¹ì¸/ì‹œì²­ ----------------
    def live_request(self):
        if self.current_user == "guest":
            messagebox.showwarning("ì œí•œ", "ê²ŒìŠ¤íŠ¸ëŠ” ë°©ì†¡ ì‹ ì²­ ë¶ˆê°€(íšŒì› ë¡œê·¸ì¸ í•„ìš”).")
            return
        title = simpledialog.askstring("ë°©ì†¡ ì‹ ì²­", "ë°©ì†¡ ì œëª©:")
        if not title:
            return
        cat = simpledialog.askstring("ë°©ì†¡ ì‹ ì²­", "ì¹´í…Œê³ ë¦¬(ì£¼ì‹/íˆ¬ì/ì¡ë‹´):") or "ì£¼ì‹"
        memo = simpledialog.askstring("ë°©ì†¡ ì‹ ì²­", "ë©”ëª¨(ì„ íƒ):") or ""
        try:
            self.repo.create_broadcast_request(self.current_user, title, category=cat, memo=memo)
            messagebox.showinfo("ì ‘ìˆ˜", "ë°©ì†¡ ì‹ ì²­ ì ‘ìˆ˜ ì™„ë£Œ!\nê´€ë¦¬ì ìŠ¹ì¸ í›„ ë°©ì†¡ ë¦¬ìŠ¤íŠ¸ì— í‘œì‹œë©ë‹ˆë‹¤.")
        except Exception as e:
            log_error_to_db(self.db, "live_request", e, {})
            apply_auto_fix(self, str(e))
            messagebox.showerror("ì˜¤ë¥˜", "ë°©ì†¡ ì‹ ì²­ ì‹¤íŒ¨(ì˜¤ë¥˜ ë¡œê·¸ ì €ì¥ë¨)")
        self.live_refresh_from_db()

    def live_refresh_from_db(self):
        if not hasattr(self, "live_tree_db"):
            return
        for r in self.live_tree_db.get_children():
            self.live_tree_db.delete(r)
        try:
            items = self.repo.list_active_broadcasts(limit=200)
        except Exception as e:
            log_error_to_db(self.db, "list_active_broadcasts", e, {})
            apply_auto_fix(self, str(e))
            items = []

        self.live_db_items = items
        for it in items:
            bid = it.get("broadcast_id") or it.get("_id") or ""
            self.live_tree_db.insert("", "end", iid=bid, values=(
                it.get("title", ""),
                it.get("category", ""),
                it.get("host", ""),
                it.get("broadcast_id", ""),
                it.get("status", ""),
                it.get("created_time", "")
            ))

    def _selected_broadcast_db(self):
        sel = self.live_tree_db.selection()
        if not sel:
            return None
        iid = sel[0]
        vals = self.live_tree_db.item(iid, "values")
        return {
            "title": vals[0],
            "category": vals[1],
            "host": vals[2],
            "broadcast_id": vals[3],
            "status": vals[4],
        }

    def live_join_selected_db(self):
        it = self._selected_broadcast_db()
        if not it:
            messagebox.showwarning("ì„ íƒ", "ë°©ì†¡ì„ ì„ íƒí•˜ì„¸ìš”.")
            return
        bid = it.get("broadcast_id") or ""
        if not bid:
            return

        # ì‹œì²­=í•´ë‹¹ ë°©ì†¡ ë°© ì±„íŒ… ì…ì¥ (MVP)
        room = f"bcast:{bid}"
        self.chat_room = room
        self.ws_send({"type": "join", "room": room})

        self._append_chat(f"[SYSTEM] ë°©ì†¡ ì…ì¥ ì™„ë£Œ: {it.get('title')} (room={room})")
        try:
            self.nb.select(self.tab_chat)
        except Exception:
            pass

    def live_end_selected_db(self):
        it = self._selected_broadcast_db()
        if not it:
            messagebox.showwarning("ì„ íƒ", "ë°©ì†¡ì„ ì„ íƒí•˜ì„¸ìš”.")
            return
        if it.get("host") != self.current_user:
            messagebox.showwarning("ê¶Œí•œ", "í˜¸ìŠ¤íŠ¸ë§Œ ì¢…ë£Œí•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.")
            return
        bid = it.get("broadcast_id") or ""
        ok, msg = self.repo.end_broadcast(bid, self.current_user)
        if ok:
            messagebox.showinfo("ì™„ë£Œ", msg)
            self.live_refresh_from_db()
        else:
            messagebox.showerror("ì˜¤ë¥˜", msg)

    def open_broadcast_admin_panel(self):
        win = tk.Toplevel(self.root)
        win.title("ê´€ë¦¬ì: ë°©ì†¡ì‹ ì²­ ìŠ¹ì¸")
        win.geometry("1200x720")
        win.configure(bg="white")

        tk.Label(win, text="ë°©ì†¡ ì‹ ì²­(PENDING) ëª©ë¡", bg="white", fg="#0064FF",
                 font=("ë§‘ì€ ê³ ë”•", 16, "bold")).pack(pady=10)

        cols = ("ìš”ì²­ID", "íšŒì›ID", "ì œëª©", "ì¹´í…Œê³ ë¦¬", "ë©”ëª¨", "ì‹œê°„")
        tree = ttk.Treeview(win, columns=cols, show="headings", height=18)
        for c in cols:
            tree.heading(c, text=c)
            tree.column(c, anchor="center")
        tree.column("ìš”ì²­ID", width=140)
        tree.column("íšŒì›ID", width=140)
        tree.column("ì œëª©", width=360)
        tree.column("ì¹´í…Œê³ ë¦¬", width=120)
        tree.column("ë©”ëª¨", width=260)
        tree.column("ì‹œê°„", width=160)
        tree.pack(fill="both", expand=True, padx=12, pady=12)

        def load():
            for r in tree.get_children():
                tree.delete(r)
            items = self.repo.list_broadcast_requests(status="PENDING", limit=200)
            for it in items:
                tree.insert("", "end", iid=it["_id"], values=(
                    it["_id"][:12],
                    it.get("user_id", ""),
                    it.get("title", ""),
                    it.get("category", ""),
                    (it.get("memo", "") or "")[:60],
                    it.get("created_time", "")
                ))

        def approve():
            sel = tree.selection()
            if not sel:
                messagebox.showwarning("ì„ íƒ", "ìŠ¹ì¸í•  ì‹ ì²­ì„ ì„ íƒí•˜ì„¸ìš”.")
                return
            req_id = sel[0]
            ok, msg = self.repo.approve_broadcast_request(req_id, self.current_user)
            if ok:
                messagebox.showinfo("ìŠ¹ì¸", msg)
                load()
                self.live_refresh_from_db()
            else:
                messagebox.showerror("ì˜¤ë¥˜", msg)

        bar = tk.Frame(win, bg="white")
        bar.pack(fill="x", padx=12, pady=(0, 12))
        tk.Button(bar, text="ìƒˆë¡œê³ ì¹¨", bg="#0064FF", fg="white",
                  font=("ë§‘ì€ ê³ ë”•", 11, "bold"), command=load).pack(side="left", padx=6)
        tk.Button(bar, text="ì„ íƒ ìŠ¹ì¸", bg="#111111", fg="white",
                  font=("ë§‘ì€ ê³ ë”•", 11, "bold"), command=approve).pack(side="right", padx=6)
        load()

    # ---------------- Chat ----------------
    def _append_chat(self, line: str):
        self.chat_text.configure(state="normal")
        self.chat_text.insert(tk.END, line + "\n")
        self.chat_text.see(tk.END)
        self.chat_text.configure(state="disabled")

    def _balloon_anim(self, count: int):
        w = max(600, self.chat_canvas.winfo_width())
        items = []
        for _ in range(max(1, min(10, count))):
            x = max(30, w // 2 + random.randint(-180, 180))
            y = 240 + random.randint(-10, 10)
            items.append(self.chat_canvas.create_text(x, y, text="ğŸˆ", font=("ë§‘ì€ ê³ ë”•", 26, "bold")))
        steps = {"n": 0}

        def step():
            steps["n"] += 1
            for it in items:
                try:
                    self.chat_canvas.move(it, 0, -10)
                except Exception:
                    pass
            if steps["n"] > 24:
                for it in items:
                    try:
                        self.chat_canvas.delete(it)
                    except Exception:
                        pass
                return
            self.chat_canvas.after(40, step)

        step()

    def chat_send(self):
        txt = self.chat_entry.get().strip()
        if not txt:
            return
        if not self._can_send_client():
            messagebox.showwarning("ì œí•œ", "ì „ì†¡ ì†ë„ ì œí•œ(2ì´ˆ)")
            return
        self.ws_send({"type": "chat", "room": self.chat_room, "user": self.current_user, "message": txt})
        self.chat_entry.delete(0, tk.END)

    def chat_send_balloon(self):
        if self.current_user == "guest":
            messagebox.showwarning("ì œí•œ", "ê²ŒìŠ¤íŠ¸ëŠ” ë³„í’ì„  ì œí•œ")
            return
        if not self._can_send_client():
            messagebox.showwarning("ì œí•œ", "ì „ì†¡ ì†ë„ ì œí•œ(2ì´ˆ)")
            return
        cnt = safe_int(simpledialog.askstring("ë³„í’ì„ ", "ëª‡ ê°œ?"), 1)
        if cnt <= 0:
            return
        total_cost = BALLOON_PRICE * cnt
        if float(self.user_data.get("balance_cash", 0.0)) < total_cost:
            messagebox.showwarning("ìºì‰¬ ë¶€ì¡±", f"CASH ë¶€ì¡±\ní•„ìš”: {total_cost:,.0f}\në³´ìœ : {self.user_data['balance_cash']:,.0f}")
            return

        self.user_data["balance_cash"] -= total_cost
        self.repo.member_update(self.current_user, {"balance_cash": self.user_data["balance_cash"]})
        self.refresh_topbar()

        self.ws_send({"type": "gift", "room": self.chat_room, "user": self.current_user, "gift": "ğŸˆë³„í’ì„ ", "count": cnt,
                      "message": f"ğŸˆ x{cnt} (CASH -{total_cost:,.0f})"})

    # ---------------- WS receive ----------------
    def _on_ws_payload(self, payload: dict):
        typ = payload.get("type")

        if typ == "system_local":
            self._append_chat(f"[LOCAL] {payload.get('message')}")
            return

        if typ == "system":
            if payload.get("room") != self.chat_room:
                return
            self._append_chat(f"[SYSTEM] {payload.get('message','')}")
            return

        if typ == "chat":
            if payload.get("room") != self.chat_room:
                return
            self._append_chat(f"{payload.get('time','')} {payload.get('user','')}: {payload.get('message','')}")
            return

        if typ == "gift":
            if payload.get("room") != self.chat_room:
                return
            cnt = int(payload.get("count", 1))
            self._append_chat("â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”")
            self._append_chat(f"{payload.get('time','')} {payload.get('user','')} â–¶ ğŸˆë³„í’ì„  x{cnt}  {payload.get('message','')}")
            self._append_chat("â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”")
            self._balloon_anim(cnt)
            return

    # ---------------- Signals / Prices ----------------
    def job_refresh_signals(self):
        out = {}
        for _, tkr in (DEFAULT_KR + DEFAULT_US):
            out[tkr] = compute_signal_simple(tkr)
        return out

    def _apply_signals(self, sig_map: dict):
        for tree in [getattr(self, "tree_kr", None), getattr(self, "tree_us", None)]:
            if not tree:
                continue
            for iid in tree.get_children():
                name, ticker, _sig = tree.item(iid, "values")
                sig = sig_map.get(ticker, "HOLD")
                tree.item(iid, values=(name, ticker, sig), tags=(sig,))

    def refresh_prices(self):
        self.work_q.put((self.job_refresh_prices, ()))

    def job_refresh_prices(self):
        prices = {}
        holdings = self.user_data.get("holdings", {}) or {}
        for tkr in holdings.keys():
            p = fetch_market_price(tkr)
            if p and p > 0:
                prices[tkr] = float(p)
        return prices

    def _apply_prices(self, prices: dict):
        self._render_portfolio(prices)
        self._render_analysis(prices)

    # ---------------- Renderers ----------------
    def _render_recent(self):
        self.recent_list.delete(0, tk.END)
        for r in self.user_data.get("recent_stocks", [])[:20]:
            self.recent_list.insert(tk.END, f"{r.get('time','')}  {r.get('name','')} ({r.get('ticker','')})")

    def _render_portfolio(self, prices: dict):
        for r in self.tree_port.get_children():
            self.tree_port.delete(r)

        holdings = self.user_data.get("holdings", {}) or {}
        for ticker, h in holdings.items():
            name = h.get("name", ticker)
            qty = float(h.get("qty", 0.0))
            avg = float(h.get("avg_price", 0.0))
            cur = float(prices.get(ticker) or 0.0)
            pnl = (cur - avg) * qty if cur > 0 else 0.0
            rr = ((cur / avg) - 1.0) * 100.0 if (cur > 0 and avg > 0) else 0.0
            self.tree_port.insert("", "end", iid=ticker, values=(
                name, ticker, f"{qty:g}", f"{avg:,.4f}", f"{cur:,.4f}" if cur else "-", f"{pnl:,.2f}", f"{rr:,.2f}"
            ))

    def _set_analysis(self, text: str):
        self.analysis_text.configure(state="normal")
        self.analysis_text.delete("1.0", tk.END)
        self.analysis_text.insert("1.0", text)
        self.analysis_text.configure(state="disabled")

    def _render_analysis(self, prices: dict):
        holdings = self.user_data.get("holdings", {}) or {}
        lines = []
        total_pnl = 0.0
        total_cost = 0.0

        for ticker, h in holdings.items():
            name = h.get("name", ticker)
            qty = float(h.get("qty", 0.0))
            avg = float(h.get("avg_price", 0.0))
            cur = float(prices.get(ticker) or 0.0)

            cost = avg * qty
            pnl = (cur - avg) * qty if cur > 0 else 0.0
            rr = ((cur / avg) - 1.0) * 100.0 if (cur > 0 and avg > 0) else 0.0

            total_pnl += pnl
            total_cost += cost
            lines.append(f"- {name}({ticker}) ìˆ˜ëŸ‰ {qty:g} | í‰ë‹¨ {avg:,.4f} | í˜„ì¬ {cur:,.4f} | ì†ìµ {pnl:,.2f} | ìˆ˜ìµë¥  {rr:,.2f}%")

        total_rr = (total_pnl / total_cost * 100.0) if total_cost > 0 else 0.0

        summary = [
            "ğŸ“Š ë³´ìœ ì¢…ëª© ìˆ˜ìµë¶„ì„",
            f"ì´ í‰ê°€ì†ìµ: {total_pnl:,.2f}",
            f"ì´ ìˆ˜ìµë¥ : {total_rr:,.2f}%",
            "",
            "ìƒì„¸:"
        ]
        summary.extend(lines if lines else ["- ë³´ìœ ì¢…ëª© ì—†ìŒ"])
        self._set_analysis("\n".join(summary))

    def _render_orders(self):
        for r in self.tree_orders.get_children():
            self.tree_orders.delete(r)
        for o in (self.orders_cache or [])[:200]:
            t = o.get("time", "")
            side = o.get("side", "")
            name = o.get("name", "")
            ticker = o.get("ticker", "")
            qty = float(o.get("qty", 0.0))
            price = float(o.get("price", 0.0))
            cur = o.get("currency", "")
            amount = float(o.get("amount", 0.0))
            self.tree_orders.insert("", "end", values=(t, side, name, ticker, f"{qty:g}", f"{price:,.4f}", cur, f"{amount:,.2f}"))

    def reload_orders(self):
        if self.current_user == "guest":
            messagebox.showinfo("ì•ˆë‚´", "ê²ŒìŠ¤íŠ¸ëŠ” ì£¼ë¬¸ë‚´ì—­ ì €ì¥ ì œí•œ")
            return
        self.orders_cache = self.repo.list_orders_for_user(self.current_user, limit=200)
        self._render_orders()
        self._refresh_kpis()

    # ---------------- Buying/Selling ----------------
    def open_market_buy_popup(self, name: str, ticker: str, sig: str):
        currency = "KRW" if (ticker.endswith(".KS") or ticker.endswith(".KQ")) else "USD"

        win = tk.Toplevel(self.root)
        win.title("ì‹œì¥ê°€ ë§¤ìˆ˜(í˜„ì¬ê°€ ìˆ˜ì • ë¶ˆê°€)")
        win.geometry("640x640")
        win.configure(bg="white")

        tk.Label(win, text=f"{name} ({ticker})", bg="white", fg="#0064FF",
                 font=("ë§‘ì€ ê³ ë”•", 16, "bold")).pack(pady=10)
        tk.Label(win, text=f"ì‹ í˜¸: {sig}", bg="white", fg="#111111",
                 font=("ë§‘ì€ ê³ ë”•", 12, "bold")).pack(pady=2)

        tk.Label(win, text="í˜„ì¬ì‹œì¥ê°€(ìˆ˜ì • ë¶ˆê°€)", bg="white", font=("ë§‘ì€ ê³ ë”•", 12, "bold")).pack(pady=(12, 4))

        price_var = tk.StringVar(value="ë¡œë”©ì¤‘...")
        price_entry = tk.Entry(win, textvariable=price_var, state="readonly",
                               readonlybackground="#f3f5f7", fg="#111111",
                               font=("ë§‘ì€ ê³ ë”•", 13), width=18)
        price_entry.pack()

        def on_price(p):
            if p and p > 0:
                price_var.set(f"{p:.6f}".rstrip("0").rstrip("."))
            else:
                price_var.set("ì¡°íšŒ ì‹¤íŒ¨")

        self.fetch_price_async(ticker, on_price, timeout_sec=4.0)

        def refresh_price():
            price_var.set("ë¡œë”©ì¤‘...")
            self.fetch_price_async(ticker, on_price, timeout_sec=4.0)

        tk.Button(win, text="í˜„ì¬ê°€ ìƒˆë¡œê³ ì¹¨", bg="#0064FF", fg="white",
                  font=("ë§‘ì€ ê³ ë”•", 10, "bold"), command=refresh_price).pack(pady=8)

        bal = tk.StringVar(value=f"ì”ê³ : KRW {self.user_data['balance_krw']:,.0f} / USD {self.user_data['balance_usd']:,.2f} / CASH {self.user_data['balance_cash']:,.0f}")
        tk.Label(win, textvariable=bal, bg="white", fg="#333333",
                 font=("ë§‘ì€ ê³ ë”•", 11, "bold")).pack(pady=8)

        frm = tk.Frame(win, bg="white")
        frm.pack(pady=10)
        tk.Label(frm, text="ìˆ˜ëŸ‰", bg="white", font=("ë§‘ì€ ê³ ë”•", 12, "bold")).grid(row=0, column=0, padx=8, pady=6, sticky="e")
        qty_entry = tk.Entry(frm, font=("ë§‘ì€ ê³ ë”•", 12), width=18)
        qty_entry.grid(row=0, column=1, padx=8, pady=6, sticky="w")

        def available_money():
            return self.user_data["balance_krw"] if currency == "KRW" else self.user_data["balance_usd"]

        def set_qty_by_pct(pct):
            price = safe_float(price_var.get(), 0.0)
            if price <= 0:
                messagebox.showwarning("ì˜¤ë¥˜", "í˜„ì¬ê°€ ì¡°íšŒ ì‹¤íŒ¨/0")
                return
            money = available_money()
            q = (money * pct) / price
            qty_entry.delete(0, tk.END)
            qty_entry.insert(0, f"{q:.6f}".rstrip("0").rstrip("."))

        btns = tk.Frame(win, bg="white")
        btns.pack(pady=8)
        tk.Button(btns, text="5%", bg="#0064FF", fg="white", font=("ë§‘ì€ ê³ ë”•", 11, "bold"),
                  width=8, command=lambda: set_qty_by_pct(0.05)).grid(row=0, column=0, padx=6, pady=6)
        tk.Button(btns, text="10%", bg="#0064FF", fg="white", font=("ë§‘ì€ ê³ ë”•", 11, "bold"),
                  width=8, command=lambda: set_qty_by_pct(0.10)).grid(row=0, column=1, padx=6, pady=6)
        tk.Button(btns, text="50%", bg="#0064FF", fg="white", font=("ë§‘ì€ ê³ ë”•", 11, "bold"),
                  width=8, command=lambda: set_qty_by_pct(0.50)).grid(row=0, column=2, padx=6, pady=6)
        tk.Button(btns, text="ì „ì•¡", bg="#111111", fg="white", font=("ë§‘ì€ ê³ ë”•", 11, "bold"),
                  width=8, command=lambda: set_qty_by_pct(1.00)).grid(row=0, column=3, padx=6, pady=6)

        def do_buy():
            # ì¡°ì‘ ë°©ì§€: í˜„ì¬ê°€ ë‹¤ì‹œ ì¡°íšŒ(ë¹„ë™ê¸° ëŒ€ì‹  ì—¬ê¸°ì„  ë¹ ë¥´ê²Œ 1íšŒ)
            price = fetch_market_price(ticker) or safe_float(price_var.get(), 0.0)
            if price <= 0:
                messagebox.showwarning("ì˜¤ë¥˜", "í˜„ì¬ê°€ ì¡°íšŒ ì‹¤íŒ¨")
                return
            price_var.set(f"{price:.6f}".rstrip("0").rstrip("."))

            qty = safe_float(qty_entry.get(), 0.0)
            if qty <= 0:
                messagebox.showwarning("ì˜¤ë¥˜", "ìˆ˜ëŸ‰ ì…ë ¥/ì„ íƒ")
                return

            cost = price * qty
            money = available_money()
            if cost > money + 1e-9:
                messagebox.showwarning("ì”ê³  ë¶€ì¡±", f"í•„ìš”: {cost:,.4f}\në³´ìœ : {money:,.4f}")
                return

            # ì”ê³  ì°¨ê°
            if currency == "KRW":
                self.user_data["balance_krw"] -= cost
            else:
                self.user_data["balance_usd"] -= cost

            # ë³´ìœ  ê°±ì‹ 
            holdings = self.user_data.get("holdings", {}) or {}
            h = holdings.get(ticker)
            if h:
                old_qty = float(h.get("qty", 0.0))
                old_avg = float(h.get("avg_price", 0.0))
                new_qty = old_qty + qty
                new_avg = ((old_avg * old_qty) + (price * qty)) / (new_qty + 1e-12)
                h["qty"] = new_qty
                h["avg_price"] = new_avg
                holdings[ticker] = h
            else:
                holdings[ticker] = {"name": name, "qty": qty, "avg_price": price, "currency": currency}
            self.user_data["holdings"] = holdings

            # ì£¼ë¬¸ ì €ì¥
            order = {
                "user_id": self.current_user,
                "side": "BUY",
                "name": name,
                "ticker": ticker,
                "qty": qty,
                "price": float(price),
                "currency": currency,
                "amount": float(cost),
                "realized_pnl": 0.0,
                "time": now_str(),
                "created_at": firestore.SERVER_TIMESTAMP
            }
            if self.current_user != "guest":
                self.repo.add_order(order)
                self.orders_cache.insert(0, order)
                self.repo.member_update(self.current_user, {
                    "balance_krw": self.user_data["balance_krw"],
                    "balance_usd": self.user_data["balance_usd"],
                    "holdings": self.user_data["holdings"]
                })

            self.refresh_topbar()
            self._render_orders()
            self.refresh_prices()
            messagebox.showinfo("ë§¤ìˆ˜ ì™„ë£Œ", f"{name}\nìˆ˜ëŸ‰: {qty}\nì²´ê²°ê°€(ì‹œì¥ê°€): {price:,.4f}")
            win.destroy()

        tk.Button(win, text="ì‹œì¥ê°€ ë§¤ìˆ˜ ì‹¤í–‰", bg="#0064FF", fg="white",
                  font=("ë§‘ì€ ê³ ë”•", 13, "bold"), width=18, height=2, command=do_buy).pack(pady=14)
        tk.Button(win, text="ë‹«ê¸°", bg="#111111", fg="white",
                  font=("ë§‘ì€ ê³ ë”•", 12, "bold"), width=18, command=win.destroy).pack(pady=6)

    def sell_selected(self):
        if self.current_user == "guest":
            messagebox.showwarning("ì œí•œ", "ê²ŒìŠ¤íŠ¸ëŠ” ë§¤ë„ ì €ì¥ ì œí•œ")
            return
        sel = self.tree_port.selection()
        if not sel:
            messagebox.showwarning("ì„ íƒ", "ë§¤ë„í•  ì¢…ëª© ì„ íƒ")
            return
        ticker = sel[0]
        h = (self.user_data.get("holdings", {}) or {}).get(ticker)
        if not h:
            return
        name = h.get("name", ticker)
        currency = h.get("currency", "KRW")
        max_qty = float(h.get("qty", 0.0))

        qty = safe_float(simpledialog.askstring("ë§¤ë„", f"{name} ë§¤ë„ ìˆ˜ëŸ‰ (ë³´ìœ  {max_qty:g})"), 0.0)
        if qty <= 0 or qty > max_qty + 1e-9:
            messagebox.showwarning("ì˜¤ë¥˜", "ìˆ˜ëŸ‰ ì˜¤ë¥˜")
            return

        price = fetch_market_price(ticker)
        if not price or price <= 0:
            messagebox.showwarning("ì˜¤ë¥˜", "í˜„ì¬ê°€ ì¡°íšŒ ì‹¤íŒ¨")
            return

        avg = float(h.get("avg_price", 0.0))
        proceeds = price * qty
        realized = (price - avg) * qty

        # ì”ê³  ì¦ê°€
        if currency == "KRW":
            self.user_data["balance_krw"] += proceeds
        else:
            self.user_data["balance_usd"] += proceeds

        # ë³´ìœ  ê°ì†Œ
        new_qty = max_qty - qty
        if new_qty <= 1e-9:
            self.user_data["holdings"].pop(ticker, None)
        else:
            h["qty"] = new_qty
            self.user_data["holdings"][ticker] = h

        order = {
            "user_id": self.current_user,
            "side": "SELL",
            "name": name,
            "ticker": ticker,
            "qty": qty,
            "price": float(price),
            "currency": currency,
            "amount": float(proceeds),
            "realized_pnl": float(realized),
            "time": now_str(),
            "created_at": firestore.SERVER_TIMESTAMP
        }
        self.repo.add_order(order)
        self.orders_cache.insert(0, order)
        self.repo.member_update(self.current_user, {
            "balance_krw": self.user_data["balance_krw"],
            "balance_usd": self.user_data["balance_usd"],
            "holdings": self.user_data["holdings"]
        })

        self.refresh_topbar()
        self._render_orders()
        self.refresh_prices()
        messagebox.showinfo("ë§¤ë„ ì™„ë£Œ", f"{name}\nìˆ˜ëŸ‰: {qty}\nì‹¤í˜„ì†ìµ: {realized:,.2f}")

    # ---------------- Charge popup ----------------
    def open_charge_popup(self):
        if self.current_user == "guest":
            messagebox.showwarning("ì œí•œ", "ê²ŒìŠ¤íŠ¸ëŠ” ì¶©ì „ ì‹ ì²­ ì œí•œ(íšŒì› ë¡œê·¸ì¸ í•„ìš”)")
            return

        win = tk.Toplevel(self.root)
        win.title("ì¶©ì „/ìœ ë£Œê²°ì œ ì‹ ì²­")
        win.geometry("760x650")
        win.configure(bg="white")

        tk.Label(win, text="ì¶©ì „í•˜ê¸° / ì´ìš©ê¶Œ ê²°ì œ(ì…ê¸ˆì‹ ì²­)", bg="white", fg="#0064FF",
                 font=("ë§‘ì€ ê³ ë”•", 18, "bold")).pack(pady=10)
        tk.Label(win, text=BOTTOM_ACCOUNT_TEXT, bg="white", fg="#666666",
                 font=("ë§‘ì€ ê³ ë”•", 12)).pack(pady=8)

        frm = tk.Frame(win, bg="white")
        frm.pack(fill="x", padx=16, pady=10)

        tk.Label(frm, text="ì…ê¸ˆìëª…", bg="white", font=("ë§‘ì€ ê³ ë”•", 12, "bold")).grid(row=0, column=0, padx=8, pady=8, sticky="e")
        depositor = tk.Entry(frm, font=("ë§‘ì€ ê³ ë”•", 12), width=28)
        depositor.grid(row=0, column=1, padx=8, pady=8, sticky="w")

        tk.Label(frm, text="íœ´ëŒ€í°", bg="white", font=("ë§‘ì€ ê³ ë”•", 12, "bold")).grid(row=1, column=0, padx=8, pady=8, sticky="e")
        phone = tk.Entry(frm, font=("ë§‘ì€ ê³ ë”•", 12), width=28)
        phone.grid(row=1, column=1, padx=8, pady=8, sticky="w")

        memo = tk.Entry(win, font=("ë§‘ì€ ê³ ë”•", 11))
        memo.pack(fill="x", padx=16, pady=(0, 10))
        memo.insert(0, "ì…ê¸ˆ í™•ì¸ í›„ ì²˜ë¦¬ ë°”ëë‹ˆë‹¤.")

        def submit_plan(plan_type: str):
            dep = depositor.get().strip()
            ph = phone.get().strip()
            if not dep or not ph:
                messagebox.showwarning("ì˜¤ë¥˜", "ì…ê¸ˆìëª…/íœ´ëŒ€í° ì…ë ¥")
                return
            if plan_type == "PLAN_1M":
                amt = PLAN_1M_PRICE
                title = f"ì´ìš©ê¶Œ 1ë‹¬({amt:,}ì›)"
            else:
                amt = PLAN_1Y_PRICE
                title = f"ì´ìš©ê¶Œ 1ë…„({amt:,}ì›)"
            self.repo.create_payment_request(self.current_user, plan_type, dep, amt, ph, memo.get().strip() + " / " + title)
            messagebox.showinfo("ì ‘ìˆ˜", f"{title} ì‹ ì²­ ì ‘ìˆ˜ ì™„ë£Œ.\nê´€ë¦¬ì ìŠ¹ì¸ í›„ ë°˜ì˜ë©ë‹ˆë‹¤.")
            win.destroy()

        def submit_wallet(kind: str):
            dep = depositor.get().strip()
            ph = phone.get().strip()
            if not dep or not ph:
                messagebox.showwarning("ì˜¤ë¥˜", "ì…ê¸ˆìëª…/íœ´ëŒ€í° ì…ë ¥")
                return
            amt = safe_float(simpledialog.askstring("ì¶©ì „ ê¸ˆì•¡", f"{kind} ì¶©ì „ ê¸ˆì•¡(ìˆ«ì):"), 0.0)
            if amt <= 0:
                messagebox.showwarning("ì˜¤ë¥˜", "ê¸ˆì•¡ ì˜¤ë¥˜")
                return
            self.repo.create_payment_request(self.current_user, kind, dep, amt, ph, memo.get().strip() + f" / {kind} ì¶©ì „")
            messagebox.showinfo("ì ‘ìˆ˜", f"{kind} ì¶©ì „ ì‹ ì²­ ì ‘ìˆ˜ ì™„ë£Œ.\nê´€ë¦¬ì ìŠ¹ì¸ í›„ ë°˜ì˜ë©ë‹ˆë‹¤.")
            win.destroy()

        box = tk.LabelFrame(win, text="ì´ìš©ê¶Œ(ìœ ë£Œê²°ì œ)", bg="white", font=("ë§‘ì€ ê³ ë”•", 12, "bold"))
        box.pack(fill="x", padx=16, pady=12)
        tk.Button(box, text=f"1ë‹¬ ì´ìš©ê¶Œ ({PLAN_1M_PRICE:,}ì›) ì‹ ì²­", bg="#0064FF", fg="white",
                  font=("ë§‘ì€ ê³ ë”•", 13, "bold"), height=2, command=lambda: submit_plan("PLAN_1M")).pack(fill="x", padx=12, pady=8)
        tk.Button(box, text=f"1ë…„ ì´ìš©ê¶Œ ({PLAN_1Y_PRICE:,}ì›) ì‹ ì²­", bg="#111111", fg="white",
                  font=("ë§‘ì€ ê³ ë”•", 13, "bold"), height=2, command=lambda: submit_plan("PLAN_1Y")).pack(fill="x", padx=12, pady=8)

        box2 = tk.LabelFrame(win, text="ì§€ê°‘ ì¶©ì „(ì„ íƒ)", bg="white", font=("ë§‘ì€ ê³ ë”•", 12, "bold"))
        box2.pack(fill="x", padx=16, pady=12)
        row = tk.Frame(box2, bg="white")
        row.pack(padx=12, pady=10)
        tk.Button(row, text="KRW", bg="#0064FF", fg="white", font=("ë§‘ì€ ê³ ë”•", 12, "bold"),
                  command=lambda: submit_wallet("KRW")).pack(side="left", padx=6)
        tk.Button(row, text="USD", bg="#0064FF", fg="white", font=("ë§‘ì€ ê³ ë”•", 12, "bold"),
                  command=lambda: submit_wallet("USD")).pack(side="left", padx=6)
        tk.Button(row, text="CASH", bg="#0064FF", fg="white", font=("ë§‘ì€ ê³ ë”•", 12, "bold"),
                  command=lambda: submit_wallet("CASH")).pack(side="left", padx=6)

        tk.Button(win, text="ë‹«ê¸°", bg="#cccccc", fg="#111111",
                  font=("ë§‘ì€ ê³ ë”•", 12, "bold"), command=win.destroy).pack(pady=10)


# =========================
# Run
# =========================
def main():
    root = tk.Tk()
    app = STApp(root)
    root.mainloop()

if __name__ == "__main__":
    try:
        main()
    except Exception as e:
        print("ì¹˜ëª…ì  ì˜¤ë¥˜:", e)
        print(traceback.format_exc())
        sys.exit(1)
```
